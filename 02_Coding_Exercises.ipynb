```python
"""4.12 (Simulation: The Tortoise and the Hare) In this problem, you’ll re-
create the classic race of the tortoise and the hare. You’ll use random-
number generation to develop a simulation of this memorable event.

Our contenders begin the race at square 1 of 70 squares. Each
square represents a position along the race course. The finish
line is at square 70. The first contender to reach or pass square
70 is rewarded with a pail of fresh carrots and lettuce. The
course weaves its way up the side of a slippery mountain, so
occasionally the contenders lose ground.
A clock ticks once per second. With each tick of the clock,
your application should adjust the position of the animals
according to the rules in the table below. Use variables to keep
track of the positions of the animals (i.e., position numbers are
1–70). Start each animal at position 1 (the “starting gate”). If
an animal slips left before square 1, move it back to square 1.

Create two functions that generate the percentages in the table
for the tortoise and the hare, respectively, by producing a
random integer i in the range 1 ≤ i ≤ 10. In the function for the
tortoise, perform a “fast plod” when 1 ≤ i ≤ 5, a “slip” when 6
≤ i ≤ 7 or a “slow plod” when 8 ≤ i ≤ 10. Use a similar
technique in the function for the hare.
Begin the race by displaying

  BANG !!!!!
  AND THEY'RE OFF !!!!!

Then, for each tick of the clock (i.e., each iteration of a loop),
display a 70-position line showing the letter "T" in the
position of the tortoise and the letter "H" in the position of the
hare. Occasionally, the contenders will land on the same
square. In this case, the tortoise bites the hare, and your
application should display "OUCH!!!" at that position. All
positions other than the "T"
, the "H" or the "OUCH!!!" (in
case of a tie) should be blank.
After each line is displayed, test for whether either animal has
reached or passed square 70. If so, display the winner and
terminate the simulation. If the tortoise wins, display
TORTOISE WINS!!! YAY!!! If the hare wins, display
Hare wins. Yuch. If both animals win on the same tick of
the clock, you may want to favor the tortoise (the “underdog”),
or you may want to display "It's a tie"
. If neither animal
wins, perform the loop again to simulate the next tick of the
clock. When you’re ready to run your application, assemble a
group of fans to watch the race. You’ll be amazed at how
involved your audience gets!
"""

"""
Description:
This program simulates “The Tortoise and the Hare” race on a 70-square track.
Each clock tick (iteration), the tortoise and hare move based on the rules
and probabilities given in the table. The race begins with both animals at
square 1. The program prints the track after each tick, showing:
- 'T' at the tortoise position
- 'H' at the hare position
- 'OUCH!!!' if they land on the same square
The program ends when either animal reaches or passes square 70 and prints
the winner (or a tie).

Input:
- None (the simulation uses random number generation).

Output:
- Two start lines:
  BANG !!!!!
  AND THEY'RE OFF !!!!!
- A 70-character track line per tick showing positions of T and H.
- A final message declaring the winner or a tie.
"""

import random


def tortoise_move():
  roll = random.randint(1, 10)

  # Fast plod (50%): 1-5
  if 1 <= roll <= 5:
    return 3
  # Slip (20%): 6-7
  elif 6 <= roll <= 7:
    return -6
  # Slow plod (30%): 8-10
  else:
    return 1


def hare_move():
  roll = random.randint(1, 10)

  # Sleep (20%): 1-2
  if 1 <= roll <= 2:
    return 0
  # Big hop (20%): 3-4
  elif 3 <= roll <= 4:
    return 9
  # Big slip (10%): 5
  elif roll == 5:
    return -12
  # Small hop (30%): 6-8
  elif 6 <= roll <= 8:
    return 1
  # Small slip (20%): 9-10
  else:
    return -2


print("BANG !!!!!")
print("AND THEY'RE OFF !!!!!")

tortoise_pos = 1
hare_pos = 1

while True:
  tortoise_pos += tortoise_move()
  hare_pos += hare_move()

  # keep positions within 1..70
  if tortoise_pos < 1:
    tortoise_pos = 1
  if hare_pos < 1:
    hare_pos = 1

  if tortoise_pos > 70:
    tortoise_pos = 70
  if hare_pos > 70:
    hare_pos = 70

  track = [' '] * 70

  if tortoise_pos == hare_pos:
    ouch = "OUCH!!!"
    start = tortoise_pos - 1
    for i in range(len(ouch)):
      if start + i < 70:
        track[start + i] = ouch[i]
  else:
    track[tortoise_pos - 1] = 'T'
    track[hare_pos - 1] = 'H'

  print(''.join(track))

  # termination checks (after printing each tick)
  tortoise_finished = tortoise_pos >= 70
  hare_finished = hare_pos >= 70

  if tortoise_finished or hare_finished:
    if tortoise_finished and hare_finished:
      print("It's a tie.")
    elif tortoise_finished:
      print("TORTOISE WINS!!! YAY!!!")
    else:
      print("Hare wins. Yuch.")
    break
```

```python
"""4.13 (Arbitrary Argument List) Calculate the product of a series of
integers that are passed to the function product, which receives an
arbitrary argument list. Test your function with several calls, each with a
different number of arguments."""

def product(*args):
  """
  Description:
  This function calculates the product of a series of integers passed
  to it as an arbitrary argument list.

  Input:
  - *args: zero or more integer values

  Output:
  - The product of all the values passed to the function
  """

  result = 1

  for value in args:
    result *= value

  return result

print(product(2, 3))
print(product(1, 2, 3, 4))
print(product(5, 10))
print(product(2, 2, 2, 2, 2))
```

```python
"""
5.8 (Sieve of Eratosthenes) A prime number is an integer greater than 1
that’s evenly divisible only by itself and 1. The Sieve of Eratosthenes is an
elegant, straightforward method of finding prime numbers. The process
for finding all primes less than 1000 is:
  1. Create a 1000-element list primes with all elements initialized
  to True. List elements with prime indices (like 2, 3, 5, 7, 11, …)
  will remain True. All other list elements will eventually be set
  to False.
  2. Starting with index 2, if a given element is True iterate through
  the rest of the list and set to False every element in primes
  whose index is a multiple of the index for the element we’re
  currently processing. For list index 2, all elements beyond
  element 2 in the list that have indices which are multiples of 2
  (i.e., 4, 6, 8, 10, …, 998) will be set to False.
  3. Repeat Step (b) for the next True element. For list index 3
  (which was initialized to True), all elements beyond element 3
  in the list that have indices which are multiples of 3 (i.e., 6, 9, 12,
  15, …, 999) will be set to False; and so on. A subtle
  observation (think about why this is true): The square root of 999
  is 31.6, you’ll need to test and set to False only all multiples of
  2, 3, 5, 7, 9, 11, 13, 17, 19, 23, 29 and 31. This will significantly
  improve the performance of your algorithm, especially if you
  decide to look for large prime numbers.
When this process completes, the list elements that are still True indicate
that the index is a prime number. These indices can be displayed. Use a list
of 1000 elements to determine and display the prime numbers less than
1000. Ignore list elements 0 and 1. [As you work through the book, you’ll

"""

"""
Description:
This program uses the Sieve of Eratosthenes to determine and display all
prime numbers less than 1000. A list of 1000 Boolean values is created,
where each index represents a number. Elements corresponding to prime
indices remain True, while non-prime indices are set to False.

Input:
- None

Output:
- All prime numbers less than 1000, displayed one per line.
"""

# create list of 1000 True values
primes = [True] * 1000

# 0 and 1 are not prime
primes[0] = False
primes[1] = False

# apply the sieve
for index in range(2, int(1000 ** 0.5) + 1):
  if primes[index]:
    for multiple in range(index * index, 1000, index):
      primes[multiple] = False

# display prime numbers
print("Prime numbers less than 1000:")

for number in range(2, 1000):
  if primes[number]:
    print(number)
```

```python
"""5.9 (Palindrome Tester) A string that’s spelled identically backward and
forward, like 'radar', is a palindrome. Write a function
is_palindrome that takes a string and returns True if it’s a
palindrome and False otherwise. Use a stack (simulated with a list as we
did in Section 5.11) to help determine whether a string is a palindrome.
Your function should ignore case sensitivity (that is, 'a' and 'A' are the
same), spaces and punctuation.
"""
"""
Description:
This function determines whether a given string is a palindrome.
A palindrome reads the same forward and backward. The function ignores
case sensitivity, spaces, and punctuation. A stack (implemented using
a list) is used to help compare characters.

Input:
- A string containing letters, spaces, and punctuation.

Output:
- True if the string is a palindrome.
- False otherwise.
"""

def is_palindrome(text):
  stack = []
  cleaned = []

  # remove non-letter characters and normalize case
  for char in text:
    if char.isalnum():
      cleaned.append(char.lower())

  # push characters onto the stack
  for char in cleaned:
    stack.append(char)

  # compare characters by popping from stack
  for char in cleaned:
    if char != stack.pop():
      return False

  return True
```

```python
"""5.10 (Anagrams) An anagram of a string is another string formed by
rearranging the letters in the first. Write a script that produces all possible
anagrams of a given string using only techniques that you’ve seen to this
point in the book. [The itertools module provides many functions,
including one that produces permutations.]"""

"""
Description:
This program produces all possible anagrams of a given string.
An anagram is formed by rearranging the letters of the original string.
The solution uses recursion, lists, and string slicing—without using
the itertools module.

Input:
- A string provided by the user.

Output:
- All possible anagrams of the string, displayed one per line.
"""

def anagrams(text):
  # base case: a single character
  if len(text) == 1:
    return [text]

  results = []

  # generate anagrams by fixing each character
  for index in range(len(text)):
    first_char = text[index]
    remaining = text[:index] + text[index + 1:]

    for sub_anagram in anagrams(remaining):
      results.append(first_char + sub_anagram)

  return results


# get input from user
word = input("Enter a word: ")

# generate and display anagrams
print("Anagrams:")
for item in anagrams(word):
  print(item)
```

```python
"""5.11 (Summarizing Letters in a String) Write a function
summarize_letters that receives a string and returns a list of tuples
containing the unique letters and their frequencies in the string. Test your
function and display each letter with its frequency. Your function should
ignore case sensitivity (that is, 'a' and 'A' are the same) and ignore
spaces and punctuation. When done, write a statement that says whether
the string has all the letters of the alphabet."""

"""
Description:
This program defines a function summarize_letters that receives a string
and returns a list of tuples containing each unique letter and its
frequency in the string. The function ignores case sensitivity, spaces,
and punctuation. After displaying each letter and its frequency, the
program determines whether the string contains all the letters of the
alphabet.

Input:
- A string containing letters, spaces, and punctuation.

Output:
- A list of tuples showing each unique letter and its frequency.
- A statement indicating whether the string contains all letters
  of the alphabet.
"""

def summarize_letters(text):
  frequencies = {}

  # count letters, ignoring case and non-letters
  for char in text:
    if char.isalpha():
      letter = char.lower()
      if letter in frequencies:
        frequencies[letter] += 1
      else:
        frequencies[letter] = 1

  # convert dictionary to list of tuples
  return list(frequencies.items())
```

```python
"""5.12 (Telephone-Number Word Generator) You should find this exercise
to be entertaining. Standard telephone keypads contain the digits zero
through nine. The numbers two through nine each have three letters
associated with them, as shown in the following table:"""

"""
Description:
This script generates all possible seven-letter word combinations for a
seven-digit telephone number using the standard phone keypad mapping:

2: ABC, 3: DEF, 4: GHI, 5: JKL, 6: MNO, 7: PQRS, 8: TUV, 9: WXYZ

The user enters a seven-digit number (digits 2–9 only). The script then
generates every possible letter combination corresponding to that number.
(Phone numbers containing 0 or 1 are rejected because they have no letters.)

Input:
- A seven-digit phone number as a string, containing only digits 2–9.

Output:
- Every possible seven-letter combination corresponding to the phone number.
- The total number of combinations generated.
"""

keypad = {
  '2': 'ABC',
  '3': 'DEF',
  '4': 'GHI',
  '5': 'JKL',
  '6': 'MNO',
  '7': 'PQRS',
  '8': 'TUV',
  '9': 'WXYZ'
}

phone = input("Enter a seven-digit phone number (digits 2–9 only): ").strip()

# validate: must be 7 digits, and each must be 2..9
valid = True

if len(phone) != 7:
  valid = False
else:
  for digit in phone:
    if digit not in keypad:
      valid = False
      break

if not valid:
  print("Invalid input. Enter exactly 7 digits using only 2–9 (no 0s or 1s).")
else:
  combinations = ['']  # start with one empty partial combination

  for digit in phone:
    letters = keypad[digit]
    next_combinations = []

    for partial in combinations:
      for letter in letters:
        next_combinations.append(partial + letter)

    combinations = next_combinations

  for word in combinations:
    print(word)

  print("\nTotal combinations:", len(combinations))
```

```python
"""6.5 (Counting Duplicate Words) Write a script that uses a dictionary to
determine and print the number of duplicate words in a sentence. Treat
uppercase and lowercase letters the same and assume there is no
punctuation in the sentence. Use the techniques you learned in Section
6.2.7. Words with counts larger than 1 have duplicates."""

"""
Description:
This script uses a dictionary to count the number of occurrences of each
word in a sentence. Uppercase and lowercase letters are treated the same,
and the sentence is assumed to contain no punctuation. After counting,
the script displays only the words that occur more than once along with
their counts.

Input:
- A sentence entered by the user (no punctuation).

Output:
- Each duplicate word and the number of times it appears in the sentence.
"""

sentence = input("Enter a sentence: ").lower()

words = sentence.split()
counts = {}

# count occurrences of each word
for word in words:
  if word in counts:
    counts[word] += 1
  else:
    counts[word] = 1

print("\nDuplicate words:")

found_duplicates = False

for word, count in counts.items():
  if count > 1:
    print(f"{word}: {count}")
    found_duplicates = True

if not found_duplicates:
  print("No duplicate words found.")
```

```python
"""6.8 (Challenge: Writing the Word Equivalent of a Check Amount) In
check-writing systems, it’s crucial to prevent alteration of check amounts.
One common security method requires that the amount be written in
numbers and spelled out in words as well. Even if someone can alter the
numerical amount of the check, it’s tough to change the amount in words.
Create a dictionary that maps numbers to their corresponding word
equivalents. Write a script that inputs a numeric check amount that’s less
than 1000 and uses the dictionary to write the word equivalent of the
amount. For example, the amount 112.43 should be written as

  ONE HUNDRED TWELVE AND 43/100
"""
"""
Description:
This script inputs a numeric check amount less than 1000 and converts the
whole-dollar portion into words using dictionaries. The cents portion is
displayed as a fraction over 100. The final output matches the standard
check-writing format.

Input:
- A numeric check amount less than 1000 (e.g., 112.43).

Output:
- The word equivalent of the check amount in uppercase letters,
  followed by the cents as a fraction (e.g., ONE HUNDRED TWELVE AND 43/100).
"""

ones = {
  0: '',
  1: 'ONE', 2: 'TWO', 3: 'THREE', 4: 'FOUR', 5: 'FIVE',
  6: 'SIX', 7: 'SEVEN', 8: 'EIGHT', 9: 'NINE',
  10: 'TEN', 11: 'ELEVEN', 12: 'TWELVE', 13: 'THIRTEEN',
  14: 'FOURTEEN', 15: 'FIFTEEN', 16: 'SIXTEEN',
  17: 'SEVENTEEN', 18: 'EIGHTEEN', 19: 'NINETEEN'
}

tens = {
  2: 'TWENTY', 3: 'THIRTY', 4: 'FORTY',
  5: 'FIFTY', 6: 'SIXTY', 7: 'SEVENTY',
  8: 'EIGHTY', 9: 'NINETY'
}

amount = float(input("Enter a check amount less than 1000: "))

dollars = int(amount)
cents = int(round((amount - dollars) * 100))

words = ''

# hundreds place
if dollars >= 100:
  words += ones[dollars // 100] + ' HUNDRED '
  dollars = dollars % 100

# tens and ones
if dollars >= 20:
  words += tens[dollars // 10] + ' '
  words += ones[dollars % 10]
else:
  words += ones[dollars]

# remove extra spaces
words = words.strip()

print(f"{words} AND {cents:02d}/100")
```

```python
"""6.11 (Analyzing the Game of Craps) Modify the script of Fig.4.2 to play
1,000,000 games of craps. Use a wins dictionary to keep track of the
number of games won for a particular number of rolls. Similarly, use a
losses dictionary to keep track of the number of games lost for a
particular number of rolls. As the simulation proceeds, keep updating the
dictionaries.
A typical key–value pair in the wins dictionary might be

  4: 50217

indicating that 50217 games were won on the 4th roll. Display a
summary of the results including:
  1. the percentage of the total games played that were won.
  2. the percentage of the total games played that were lost.
  3. the percentages of the total games played that were won or lost
  on a given roll (column 2 of the sample output).
  4. the cumulative percentage of the total games played that were
  won or lost up to and including a given number of rolls (column
  3 of the sample output).
Your output should be similar to the one below.
"""

"""
Description:
This script simulates 1,000,000 games of craps (based on Deitel Fig. 4.2).
It uses two dictionaries:
- wins[roll_count] = number of games won on that roll
- losses[roll_count] = number of games lost on that roll

After the simulation, it displays:
1) Percent of total games won
2) Percent of total games lost
3) Percent of total games decided (won/lost) on each roll
4) Cumulative percent of total games decided up to each roll

Input:
- None

Output:
- Overall win/loss percentages
- A table of roll-by-roll decision percentages and cumulative percentages
"""

import random


def roll_dice():
  die1 = random.randrange(1, 7)
  die2 = random.randrange(1, 7)
  return die1 + die2


GAMES = 1_000_000

wins = {}
losses = {}

for _ in range(GAMES):
  roll_count = 1
  sum_of_dice = roll_dice()

  # first roll rules
  if sum_of_dice in (7, 11):
    wins[roll_count] = wins.get(roll_count, 0) + 1
    continue
  elif sum_of_dice in (2, 3, 12):
    losses[roll_count] = losses.get(roll_count, 0) + 1
    continue

  point = sum_of_dice

  # subsequent rolls
  while True:
    roll_count += 1
    sum_of_dice = roll_dice()

    if sum_of_dice == point:
      wins[roll_count] = wins.get(roll_count, 0) + 1
      break
    elif sum_of_dice == 7:
      losses[roll_count] = losses.get(roll_count, 0) + 1
      break


total_wins = sum(wins.values())
total_losses = sum(losses.values())

win_pct = (total_wins / GAMES) * 100
loss_pct = (total_losses / GAMES) * 100

print(f"Games played: {GAMES:,}")
print(f"Wins:  {total_wins:,} ({win_pct:.2f}%)")
print(f"Losses:{total_losses:,} ({loss_pct:.2f}%)\n")

max_roll = 1
if wins:
  max_roll = max(max_roll, max(wins))
if losses:
  max_roll = max(max_roll, max(losses))

print(f"{'Roll':>4}  {'% Decided':>10}  {'Cumulative %':>12}")
print(f"{'-' * 4}  {'-' * 10}  {'-' * 12}")

cumulative_pct = 0.0

for roll in range(1, max_roll + 1):
  decided_on_roll = wins.get(roll, 0) + losses.get(roll, 0)
  roll_pct = (decided_on_roll / GAMES) * 100
  cumulative_pct += roll_pct
  print(f"{roll:>4}  {roll_pct:>10.2f}  {cumulative_pct:>12.2f}")
```

```python
"""6.14 (Intro to Data Science: Dynamic Visualization of Coin Tossing)
Modify your coin-tossing simulation from Exercise 5.31 to update the bar
plot dynamically as you flip the coin. Use the techniques you learned in
Section 6.4.2."""

"""
Description:
This script simulates repeated coin tosses and dynamically updates a bar
plot showing the number of heads and tails after each toss. The plot is
updated in real time using matplotlib, demonstrating dynamic visualization
techniques.

Input:
- None

Output:
- A dynamically updating bar chart displaying counts of Heads and Tails.
"""

import random
import matplotlib.pyplot as plt

# enable interactive mode
plt.ion()

heads = 0
tails = 0

labels = ['Heads', 'Tails']
counts = [heads, tails]

figure, axes = plt.subplots()
bars = axes.bar(labels, counts)

axes.set_ylim(0, 100)
axes.set_title("Dynamic Coin Toss Simulation")
axes.set_ylabel("Count")

# simulate coin tosses
for toss in range(1, 101):
  coin = random.choice(['H', 'T'])

  if coin == 'H':
    heads += 1
  else:
    tails += 1

  counts[0] = heads
  counts[1] = tails

  # update bar heights
  for bar, count in zip(bars, counts):
    bar.set_height(count)

  axes.set_ylim(0, max(counts) + 5)

  plt.pause(0.05)

# keep the final plot visible
plt.ioff()
plt.show()
```

```python
"""6.16 (Intro to Data Science: Dynamic Visualization of the Dice Game of
Craps) Reimplement your solution to Exercise 5.33, using the techniques
you learned in Section 6.4.2 to create a dynamic bar chart showing the
wins and losses on the first roll, second roll, third roll, etc."""

"""
Description:
This script simulates the dice game of craps and dynamically updates a bar
chart that shows the number of wins and losses that occur on each roll.
The bar chart is updated after each game using matplotlib's interactive
mode, illustrating how outcomes are distributed across roll counts.

Input:
- None

Output:
- A dynamically updating bar chart showing wins and losses by roll number.
"""

import random
import matplotlib.pyplot as plt


def roll_dice():
  return random.randrange(1, 7) + random.randrange(1, 7)


# enable interactive plotting
plt.ion()

wins = {}
losses = {}

max_rolls_displayed = 15
rolls = list(range(1, max_rolls_displayed + 1))

win_counts = [0] * max_rolls_displayed
loss_counts = [0] * max_rolls_displayed

figure, axes = plt.subplots()
bars_win = axes.bar(rolls, win_counts, label='Wins')
bars_loss = axes.bar(rolls, loss_counts, bottom=win_counts, label='Losses')

axes.set_title("Dynamic Craps Simulation: Wins and Losses by Roll")
axes.set_xlabel("Roll Number")
axes.set_ylabel("Count")
axes.legend()
axes.set_ylim(0, 50)


GAMES = 500

for _ in range(GAMES):
  roll_count = 1
  sum_of_dice = roll_dice()

  # first roll rules
  if sum_of_dice in (7, 11):
    wins[roll_count] = wins.get(roll_count, 0) + 1
  elif sum_of_dice in (2, 3, 12):
    losses[roll_count] = losses.get(roll_count, 0) + 1
  else:
    point = sum_of_dice
    while True:
      roll_count += 1
      sum_of_dice = roll_dice()

      if sum_of_dice == point:
        wins[roll_count] = wins.get(roll_count, 0) + 1
        break
      elif sum_of_dice == 7:
        losses[roll_count] = losses.get(roll_count, 0) + 1
        break

  # update bar data (cap display at max_rolls_displayed)
  for i in range(max_rolls_displayed):
    win_counts[i] = wins.get(i + 1, 0)
    loss_counts[i] = losses.get(i + 1, 0)

  for bar, count in zip(bars_win, win_counts):
    bar.set_height(count)

  for bar, w, l in zip(bars_loss, win_counts, loss_counts):
    bar.set_height(l)
    bar.set_y(w)

  axes.set_ylim(0, max(win_counts[i] + loss_counts[i] for i in range(max_rolls_displayed)) + 5)
  plt.pause(0.03)

# keep final plot visible
plt.ioff()
plt.show()
```
