```python
"""7.25 (Knight’s Tour Project: Brute-Force Approaches) In Part (c) of the
previous exercise, we developed a solution to the Knight’s Tour problem.
The approach used, called the “accessibility heuristic,” generates many
solutions and executes efficiently.
As computers continue to increase in power, we’ll be able to solve more
problems with sheer computer power and relatively unsophisticated
algorithms. Let’s call this approach “brute-force” problem solving.
1. Use random-number generation to enable the knight to walk
around the chessboard (in its legitimate L-shaped moves) at
random. Your script should run one tour and display the final
chessboard. How far did the knight get?
2. Most likely, the script in Part (a) produced a relatively short tour.
Now modify your script to attempt 1,000,000 tours. Use a onedimensional
array to keep track of the number of tours of each
length. When your script finishes attempting the 1,000,000 tours,
it should display this information in a neat tabular format. What
was the best result?
3. Most likely, the script in Part (b) gave you some “respectable”
tours, but no full tours. Now let your script run until it produces a
full tour. [Caution: This version of the script could run for hours
on a powerful computer.] Once again, keep a table of the number
of tours of each length, and display this table when the first full
tour is found. How many tours did your script attempt before
producing a full tour? How much time did it take?
4. Compare the brute-force version of the Knight’s Tour with the
accessibility-heuristic version. Which required a more careful
study of the problem? Which algorithm was more challenging to
develop? Which required more computer power? Could we be
certain (in advance) of obtaining a full tour with the accessibilityheuristic
approach? Could we be certain (in advance) of
obtaining a full tour with the brute-force approach? Argue the
pros and cons of brute-force problem-solving in general."""


# One random tour
import numpy as np
import random

BOARD_SIZE = 8

horizontal = [2, 1, -1, -2, -2, -1, 1, 2]
vertical   = [-1, -2, -2, -1, 1, 2, 2, 1]


def in_bounds(r, c):
  return 0 <= r < BOARD_SIZE and 0 <= c < BOARD_SIZE


def random_knights_tour(start_row, start_col):
  board = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=int)

  row, col = start_row, start_col
  move_count = 1
  board[row, col] = move_count

  while True:
    possible_moves = []

    for move in range(8):
      r = row + vertical[move]
      c = col + horizontal[move]
      if in_bounds(r, c) and board[r, c] == 0:
        possible_moves.append((r, c))

    if not possible_moves:
      break

    row, col = random.choice(possible_moves)
    move_count += 1
    board[row, col] = move_count

  return board, move_count
```

```python
# 1,000,000 tours

def brute_force_statistics(trials=1_000_000):
  counts = np.zeros(65, dtype=int)

  for _ in range(trials):
    _, moves = random_knights_tour(0, 0)
    counts[moves] += 1

  print("Moves  Tours")
  print("------------")
  for moves in range(1, 65):
    if counts[moves] > 0:
      print(f"{moves:>5}  {counts[moves]}")
```

```python
# Stop on first full tour

import time

def find_full_tour():
  counts = np.zeros(65, dtype=int)
  attempts = 0
  start_time = time.time()

  while True:
    attempts += 1
    _, moves = random_knights_tour(0, 0)
    counts[moves] += 1

    if moves == 64:
      elapsed = time.time() - start_time
      print("FULL TOUR FOUND!")
      print("Attempts:", attempts)
      print("Time (seconds):", elapsed)

      print("\nMove Length Distribution:")
      for i in range(1, 65):
        if counts[i] > 0:
          print(f"{i:>5}  {counts[i]}")
      break
```
