9.1 (Class Average: Writing Grades to a Plain Text File) Figure 3.2
presented a class-average script in which you could enter any number of
grades followed by a sentinel value, then calculate the class average.
Another approach would be to read the grades from a file. In an IPython
session, write code that enables you to store any number of grades into a
grades.txt plain text file.

````python
"""
Description:
Prompts the user to enter any number of grades, then writes each grade
to a plain text file named grades.txt (one grade per line).
The sentinel value -1 ends input.
"""

with open("grades.txt", "w") as file:
  while True:
    entry = input("Enter grade (-1 to end): ").strip()
    if entry == "-1":
      break
    file.write(entry + "\n")

print("Saved grades to grades.txt")
````

9.2 (Class Average: Reading Grades from a Plain Text File) In an IPython
session, write code that reads the grades from the grades.txt file you
created in the previous exercise. Display the individual grades and their
total, count and average.

````python
"""
Description:
Reads grades from grades.txt, displays each grade and then displays
the total, count, and average.
"""

total = 0
count = 0

with open("grades.txt", "r") as file:
  for line in file:
    grade = int(line.strip())
    print("Grade:", grade)
    total += grade
    count += 1

print("Total:", total)
print("Count:", count)

if count > 0:
  print("Average:", total / count)
else:
  print("Average: 0 (no grades)")
````

9.5 (Class Average: Creating a Grade Report from a CSV File) Modify
your solution to the preceding exercise to create a grade report that
displays each student’s average to the right of that student’s row and the
class average for each exam below that exam’s column.

````python
"""
Description:
Reads grades.csv and displays a grade report that includes each student's
average and the class average for each exam.
"""

import csv

records = []

with open("grades.csv", "r", newline="") as file:
  reader = csv.reader(file)
  for row in reader:
    first, last, e1, e2, e3 = row
    e1 = int(e1)
    e2 = int(e2)
    e3 = int(e3)
    records.append((first, last, e1, e2, e3))

print(f"{'First':<12}{'Last':<12}{'Exam1':>7}{'Exam2':>7}{'Exam3':>7}{'Avg':>8}")
print("-" * 53)

exam1_total = 0
exam2_total = 0
exam3_total = 0

for first, last, e1, e2, e3 in records:
  avg = (e1 + e2 + e3) / 3
  exam1_total += e1
  exam2_total += e2
  exam3_total += e3
  print(f"{first:<12}{last:<12}{e1:>7}{e2:>7}{e3:>7}{avg:>8.2f}")

count = len(records)
if count > 0:
  print("-" * 53)
  print(f"{'Class Avg':<24}{exam1_total / count:>7.2f}{exam2_total / count:>7.2f}{exam3_total / count:>7.2f}")
````

9.9 (Telephone-Number Word Generator) In Exercise 5.12, you created a
telephone-number word-generator program. Modify that program to write
its output to a text file.

````python
"""
Description:
Generates all 7-letter combinations for a 7-digit phone number (no 0 or 1)
and writes the results to phone_words.txt (one per line).
"""

digit_to_letters = {
  "2": "ABC",
  "3": "DEF",
  "4": "GHI",
  "5": "JKL",
  "6": "MNO",
  "7": "PRS",
  "8": "TUV",
  "9": "WXY"
}

number = input("Enter a 7-digit phone number (digits 2-9 only): ").strip()

valid = True
if len(number) != 7:
  valid = False
else:
  for ch in number:
    if ch not in digit_to_letters:
      valid = False
      break

if not valid:
  print("Invalid number. Use exactly 7 digits and only digits 2-9.")
else:
  results = [""]

  for digit in number:
    letters = digit_to_letters[digit]
    new_results = []
    for prefix in results:
      for letter in letters:
        new_results.append(prefix + letter)
    results = new_results

  with open("phone_words.txt", "w") as file:
    for word in results:
      file.write(word + "\n")

  print("Wrote", len(results), "combinations to phone_words.txt")
````

````python
"""
Description:
Reads a State-of-the-Union speech from a text file and displays:
- total word count
- total character count
- average word length
- average sentence length (words per sentence)
- word frequency distribution (top words + frequency-of-frequencies)
- top 10 longest words
"""

import re
from collections import Counter

filename = input("Enter the speech filename (e.g., SOTU.txt): ").strip()

with open(filename, "r", encoding="utf-8", errors="ignore") as file:
  text = file.read()

# Total character count (includes spaces/newlines)
total_characters = len(text)

# Words: letters + apostrophes (e.g., don't)
words = re.findall(r"[A-Za-z']+", text.lower())
total_words = len(words)

# Average word length
letters_only = [re.sub(r"[^a-z]", "", w) for w in words]
letters_only = [w for w in letters_only if w != ""]

total_letters = sum(len(w) for w in letters_only)
avg_word_length = (total_letters / len(letters_only)) if len(letters_only) > 0 else 0

# Sentences: split on ., !, ?
sentences = re.split(r"[.!?]+", text)
sentences = [s.strip() for s in sentences if s.strip() != ""]

# Average sentence length (words per sentence)
sentence_word_counts = []
for s in sentences:
  s_words = re.findall(r"[A-Za-z']+", s)
  sentence_word_counts.append(len(s_words))

avg_sentence_length = (sum(sentence_word_counts) / len(sentence_word_counts)) if len(sentence_word_counts) > 0 else 0

# Word frequency count
word_counts = Counter(letters_only)

# Top 10 most common words
top_10_common = word_counts.most_common(10)

# Frequency-of-frequencies:
# Example: {1: 1200, 2: 300, 3: 120, ...}
freq_of_freq = Counter(word_counts.values())

# Top 10 longest words (unique)
unique_words = set(letters_only)
longest_10 = sorted(unique_words, key=lambda w: (-len(w), w))[:10]

# -----------------------------
# Display results
# -----------------------------
print("\n--- Speech Statistics ---")
print("Total characters:", total_characters)
print("Total words:", total_words)
print("Average word length:", f"{avg_word_length:.2f}")
print("Average sentence length (words):", f"{avg_sentence_length:.2f}")

print("\n--- Top 10 Most Common Words ---")
for word, count in top_10_common:
  print(f"{word:<15}{count}")

print("\n--- Word Frequency Distribution (count -> how many words) ---")
for freq in sorted(freq_of_freq):
  print(f"{freq:>5} -> {freq_of_freq[freq]}")

print("\n--- Top 10 Longest Words (unique) ---")
for w in longest_10:
  print(f"{w:<25}({len(w)})")
````

them.

````python
"""
Description:
Account class with read-only properties for name and balance.
"""

class Account:
  def __init__(self, name, balance):
    self._name = name
    self._balance = balance

  @property
  def name(self):
    return self._name

  @property
  def balance(self):
    return self._balance

  def deposit(self, amount):
    if amount < 0:
      raise ValueError("Deposit amount must be non-negative.")
    self._balance += amount

  def withdraw(self, amount):
    if amount < 0:
      raise ValueError("Withdraw amount must be non-negative.")
    if amount > self._balance:
      raise ValueError("Insufficient funds.")
    self._balance -= amount

# Testing
acct = Account("Maxine", 100.0)
print(acct.name, acct.balance)

try:
  acct.name = "New Name"
except AttributeError as e:
  print("Name is read-only:", e)

try:
  acct.balance = 999
except AttributeError as e:
  print("Balance is read-only:", e)
````

10.4 (Modifying the Internal Data Representation of a Class) Section
10.4.2’s Time class represents the time as three integer values. Modify the
class to store the time as the total number of seconds since midnight.
Replace the _hour, _minute and _second attributes with one
_total_seconds attribute. Modify the bodies of the hour, minute
and second properties’ methods to get and set _total_seconds. Reexecute
Section 10.4’s IPython session using the modified Time class to
show that the updated class Time is interchangeable with the original one.

````python
"""
Description:
Time class with _total_seconds attribute.

The decorators (@) in the code converts a method (like hour(self))
into a read-only attribute. This makes it possible to access it without
parentheses
"""

class Time:
  def __init__(self, hour=0, minute=0, second=0):
    self._total_seconds = 0
    self.set_time(hour, minute, second)

  def set_time(self, hour, minute, second):
    if not (0 <= hour <= 23):
      raise ValueError("hour must be 0..23")
    if not (0 <= minute <= 59):
      raise ValueError("minute must be 0..59")
    if not (0 <= second <= 59):
      raise ValueError("second must be 0..59")
    self._total_seconds = hour * 3600 + minute * 60 + second

  @property
  def hour(self):
    return self._total_seconds // 3600

  @hour.setter
  def hour(self, value):
    self.set_time(value, self.minute, self.second)

  @property
  def minute(self):
    return (self._total_seconds % 3600) // 60

  @minute.setter
  def minute(self, value):
    self.set_time(self.hour, value, self.second)

  @property
  def second(self):
    return self._total_seconds % 60

  @second.setter
  def second(self, value):
    self.set_time(self.hour, self.minute, value)

  @property
  def universal_str(self):
    return f"{self.hour:02d}:{self.minute:02d}:{self.second:02d}"

  def __repr__(self):
    return f"Time({self.universal_str})"


t = Time(23, 59, 58)
print(t, t.hour, t.minute, t.second)
t.second = 10
print(t, t.hour, t.minute, t.second)
````

10.5 (Duck Typing) Recall that with duck typing, objects of unrelated
classes can respond to the same method calls if they implement those
methods. In Section 10.8, you created a list containing a
CommissionEmployee and a SalariedCommissionEmployee.
Then, you iterated through it, displaying each employee’s string
representation and earnings. Create a class SalariedEmployee for an
employee that gets paid a fixed weekly salary. Do not inherit from
CommissionEmployee or SalariedCommissionEmployee. In
class SalariedEmployee, override method __repr__ and provide
an earnings method. Demonstrate duck typing by creating an object of
your class, adding it to the list at the end of Section 10.8, then executing
the loop to show that it properly processes objects of all three classes.

````python
"""
Description:
SalariedEmployee class for an employee that gets paid a fixed weekly salary.
"""

class CommissionEmployee:
  def __init__(self, first, last, gross_sales, commission_rate):
    self.first = first
    self.last = last
    self.gross_sales = gross_sales
    self.commission_rate = commission_rate

  def earnings(self):
    return self.gross_sales * self.commission_rate

  def __repr__(self):
    return (f"CommissionEmployee({self.first} {self.last}, "
            f"sales={self.gross_sales}, rate={self.commission_rate})")


class SalariedCommissionEmployee:
  def __init__(self, first, last, gross_sales, commission_rate, base_salary):
    self.first = first
    self.last = last
    self.gross_sales = gross_sales
    self.commission_rate = commission_rate
    self.base_salary = base_salary

  def earnings(self):
    return self.base_salary + (self.gross_sales * self.commission_rate)

  def __repr__(self):
    return (f"SalariedCommissionEmployee({self.first} {self.last}, "
            f"base={self.base_salary}, sales={self.gross_sales}, rate={self.commission_rate})")


class SalariedEmployee:
  def __init__(self, first, last, weekly_salary):
    self.first = first
    self.last = last
    self.weekly_salary = weekly_salary

  def earnings(self):
    return self.weekly_salary

  def __repr__(self):
    return f"SalariedEmployee({self.first} {self.last}, weekly_salary={self.weekly_salary})"


employees = [
  CommissionEmployee("Ava", "Jones", 5000, 0.10),
  SalariedCommissionEmployee("Ben", "Smith", 3000, 0.08, 400),
  SalariedEmployee("Cara", "Lee", 950)
]

for emp in employees:
  print(emp)
  print("Earnings:", emp.earnings())
  print()
````

10.6 (Composition: A Circle “Has a” Point at Its Center) A circle has
a point at its center. Create a class Point that represents an (x-y)
coordinate pair and provides x and y read-write properties for the
attributes _x and _y. Include __init__ and __repr__ methods, and a
move method that receives x- and y-coordinate values and sets the
Point’s new location. Create a class Circle that has as its attributes
_radius and _point (a Point that represents the Circle’s center
location). Include __init__ and __repr__ methods, and a move
method that receives x- and y-coordinate values and sets a new location for
the Circle by calling the composed Point object’s move method. Test
your Circle class by creating a Circle object, displaying its string
representation, moving the Circle and displaying its string
representation again.

````python
class Point:
  def __init__(self, x=0.0, y=0.0):
    self.x = x
    self.y = y

  @property
  def x(self):
    return self._x

  @x.setter
  def x(self, value):
    self._x = float(value)

  @property
  def y(self):
    return self._y

  @y.setter
  def y(self, value):
    self._y = float(value)

  def move(self, new_x, new_y):
    self.x = new_x
    self.y = new_y

  def __repr__(self):
    return f"Point(x={self.x}, y={self.y})"

class Circle:
  def __init__(self, radius, x=0.0, y=0.0):
    self.radius = radius
    self._point = Point(x, y)

  @property
  def radius(self):
    return self._radius

  @radius.setter
  def radius(self, value):
    value = float(value)
    if value < 0:
      raise ValueError("radius must be non-negative")
    self._radius = value

  def move(self, new_x, new_y):
    self._point.move(new_x, new_y)

  def __repr__(self):
    return f"Circle(radius={self.radius}, center={self._point})"


c = Circle(5, 1, 2)
print(c)
c.move(10, 20)
print(c)
````

10.9 (Square Class) Write a class that implements a Square shape. The
class should contain a side property. Provide an __init__ method that
takes the side length as an argument. Also, provide the following readonly
properties:
```
1. perimeter returns 4 × side.
2. area returns side × side.
3. diagonal returns the square root of the expression (2 ×
side ).
```
The perimeter, area and diagonal should not have corresponding
data attributes; rather, they should use side in calculations that return the
desired values. Create a Square object and display its side,
perimeter, area and diagonal properties’ values.

````python
import math

class Square:
  def __init__(self, side):
    self.side = side

  @property
  def side(self):
    return self._side

  @side.setter
  def side(self, value):
    value = float(value)
    if value < 0:
      raise ValueError("side must be non-negative")
    self._side = value

  @property
  def perimeter(self):
    return 4 * self.side

  @property
  def area(self):
    return self.side * self.side

  @property
  def diagonal(self):
    return math.sqrt(2 * (self.side ** 2))


sq = Square(5)
print("side:", sq.side)
print("perimeter:", sq.perimeter)
print("area:", sq.area)
print("diagonal:", sq.diagonal)
````

10.10 (Invoice Class) Create a class called Invoice that a hardware
store might use to represent an invoice for an item sold at the store. An
Invoice should include four pieces of information as data attributes—a
part number (a string), a part description (a string), a quantity of the item
being purchased (an int) and a price per item (a Decimal). Your class
2
should have an __init__ method that initializes the four data attributes.
Provide a property for each data attribute. The quantity and price per item
should each be non-negative—use validation in the properties for these
data attributes to ensure that they remain valid. Provide a
calculate_invoice method that returns the invoice amount (that is,
multiplies the quantity by the price per item). Demonstrate class
Invoice’s capabilities.

````python
from decimal import Decimal

class Invoice:
  def __init__(self, part_number, part_description, quantity, price_per_item):
    self.part_number = part_number
    self.part_description = part_description
    self.quantity = quantity
    self.price_per_item = price_per_item

  @property
  def part_number(self):
    return self._part_number

  @part_number.setter
  def part_number(self, value):
    self._part_number = str(value)

  @property
  def part_description(self):
    return self._part_description

  @part_description.setter
  def part_description(self, value):
    self._part_description = str(value)

  @property
  def quantity(self):
    return self._quantity

  @quantity.setter
  def quantity(self, value):
    value = int(value)
    if value < 0:
      raise ValueError("quantity must be non-negative")
    self._quantity = value

  @property
  def price_per_item(self):
    return self._price_per_item

  @price_per_item.setter
  def price_per_item(self, value):
    value = Decimal(value)
    if value < 0:
      raise ValueError("price_per_item must be non-negative")
    self._price_per_item = value

  def calculate_invoice(self):
    return self.quantity * self.price_per_item

  def __repr__(self):
    return (f"Invoice(part_number={self.part_number}, "
            f"description={self.part_description}, "
            f"quantity={self.quantity}, price_per_item={self.price_per_item})")


inv = Invoice("A100", "Hammer", 3, "19.99")
print(inv)
print("Invoice amount:", inv.calculate_invoice())
````
