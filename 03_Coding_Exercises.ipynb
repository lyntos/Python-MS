```python
"""7.7 (Challenge Project: Reimplement NumPy array Output) You saw
that NumPy outputs two-dimensional arrays in a nice column-based
format that right-aligns every element in a field width. The field width’s
size is determined by the array element value that requires the most
character positions to display. To understand how powerful it is to have
this formatting simply built-in, write a function that reimplements
NumPy’s array formatting for two-dimensional arrays using loops.
Assume the array contains only positive integer values."""

"""
Description:
The function displays a two-dimensional list of positive integers in a
column-based format similar to NumPy's array output.

Input:
- A two-dimensional list containing only positive integers.

Output:
- The formatted array displayed with right-aligned columns.
"""

def print_2d_array(array):
  max_width = 0
  for row in array:
    for value in row:
      width = len(str(value))
      if width > max_width:
        max_width = width

  for row in array:
    for value in row:
      print(f"{value:>{max_width}}", end=" ")
    print()

#
data = [[1, 20, 300],[4000, 50, 6],[7, 800, 90,]]
print_2d_array(data)
```

```python
"""7.8 (Challenge Project: Reimplement DataFrame Output) You saw that
pandas displays DataFrames in an attractive column-based format with
row and column labels. The values within each column are right aligned in
the same field width, which is determined by that column’s widest value.
To understand how powerful it is to have this formatting built-in, write a
function that reimplements DataFrame formatting using loops. Assume
the DataFrame contains only positive integer values and that both the
row and column labels are each integer values beginning at 0."""

"""
Description:
This function displays a two-dimensional list of positive integers in a
DataFrame-like format similar to pandas. The output includes row and
column labels, and each column is right-aligned using a field width
determined by the widest value in that column.

Input:
- A two-dimensional list (rows x columns) containing positive integers.

Output:
- A formatted table with row and column labels and aligned columns.
"""

def print_dataframe(data):
  rows = len(data)
  cols = len(data[0]) # assuming all lists have same length

  # determine width of row labels
  row_label_width = len(str(rows - 1))

  # determine width for each column
  col_widths = []
  for col in range(cols):
    max_width = len(str(col))  # include column label width
    for row in range(rows):
      value_width = len(str(data[row][col]))
      if value_width > max_width:
        max_width = value_width
    col_widths.append(max_width)

  # print column headers
  print(' ' * (row_label_width + 2), end='')
  for col in range(cols):
    print(f"{col:>{col_widths[col]}} ", end='')
  print()

  # print rows with row labels
  for row in range(rows):
    print(f"{row:>{row_label_width}}  ", end='')
    for col in range(cols):
      print(f"{data[row][col]:>{col_widths[col]}} ", end='')
    print()

data = [
  [1, 20, 300],
  [4000, 50, 6],
  [7, 800, 90]
]

print_dataframe(data)
```

```python
"""7.10 (Project: Two-Player, Two-Dimensional Tic-Tac-Toe) Write a script
to play two-dimensional Tic-Tac-Toe between two human players who
alternate entering their moves on the same computer. Use a 3-by-3 twodimensional
array. Each player indicates their moves by entering a pair
of numbers representing the row and column indices of the square in
which they want to place their mark, either an 'X' or an 'O'. When the
first player moves, place an 'X' in the specified square. When the second
player moves, place an 'O' in the specified square. Each move must be to
an empty square. After each move, determine whether the game has been
won and whether it’s a draw."""

"""
Description:
This script plays two-player Tic-Tac-Toe on a 3-by-3 two-dimensional array.
Two human players alternate turns on the same computer. Player 1 places
'X' and Player 2 places 'O'. Each move is entered as a pair of integers
(row and column indices). Moves must be made to an empty square.
After each move, the script determines whether the game has been won
or whether the game is a draw.

Input:
- Row and column indices (0–2) entered by the players.

Output:
- The board after each move.
- A message indicating a win or a draw.
"""

import numpy as np


def display_board(board):
  print()
  for row in range(3):
    print(" ", end="")
    for col in range(3):
      print(board[row, col], end="")
      if col < 2:
        print(" | ", end="")
    print()
    if row < 2:
      print("---+---+---")
  print()


def is_winner(board, mark):
  # check rows and columns
  for i in range(3):
    if np.all(board[i, :] == mark):
      return True
    if np.all(board[:, i] == mark):
      return True

  # check diagonals
  if np.all(np.diag(board) == mark):
    return True
  if np.all(np.diag(np.fliplr(board)) == mark):
    return True

  return False


def is_draw(board):
  return not np.any(board == ' ')


board = np.full((3, 3), ' ', dtype='<U1')

current_player = 1
current_mark = 'X'

print("Two-Player Tic-Tac-Toe")
print("Enter moves as row and column indices from 0 to 2.")
display_board(board)

while True:
  move = input(f"Player {current_player} ({current_mark}) enter row,col: ").strip()

  # basic input validation
  if ',' not in move:
    print("Invalid format. Enter as row,col (example: 1,2).")
    continue

  parts = move.split(',')
  if len(parts) != 2:
    print("Invalid format. Enter as row,col (example: 1,2).")
    continue

  if not parts[0].strip().isdigit() or not parts[1].strip().isdigit():
    print("Row and column must be integers from 0 to 2.")
    continue

  row = int(parts[0].strip())
  col = int(parts[1].strip())

  if row < 0 or row > 2 or col < 0 or col > 2:
    print("Row and column must be in the range 0 to 2.")
    continue

  if board[row, col] != ' ':
    print("That square is not empty. Try again.")
    continue

  # place the mark
  board[row, col] = current_mark
  display_board(board)

  # check win
  if is_winner(board, current_mark):
    print(f"Player {current_player} wins!")
    break

  # check draw
  if is_draw(board):
    print("It's a draw!")
    break

  # switch players
  if current_player == 1:
    current_player = 2
    current_mark = 'O'
  else:
    current_player = 1
    current_mark = 'X'
```

```python
"""7.11 (Challenge Project: Tic-Tac-Toe with Player Against the Computer)
Modify your script from the previous exercise so that the computer makes
the moves for one of the players. Also, allow the player to specify whether
he or she wants to go first or second."""

"""
Description:
This script plays two-dimensional Tic-Tac-Toe between a human player and
the computer. The player can choose whether to go first or second.
The game uses a 3-by-3 two-dimensional array. The human player uses 'X'
and the computer uses 'O'. After each move, the script checks for a win
or a draw.

Input:
- Player choice to go first or second.
- Row and column indices for the player's moves.

Output:
- The game board after each move.
- A message indicating whether the player wins, the computer wins, or
  the game ends in a draw.
"""

import numpy as np
import random


def display_board(board):
  print()
  for row in range(3):
    print(" ", end="")
    for col in range(3):
      print(board[row, col], end="")
      if col < 2:
        print(" | ", end="")
    print()
    if row < 2:
      print("---+---+---")
  print()


def is_winner(board, mark):
  for i in range(3):
    if np.all(board[i, :] == mark):
      return True
    if np.all(board[:, i] == mark):
      return True

  if np.all(np.diag(board) == mark):
    return True
  if np.all(np.diag(np.fliplr(board)) == mark):
    return True

  return False


def is_draw(board):
  return not np.any(board == ' ')


def computer_move(board):
  empty_squares = []

  for row in range(3):
    for col in range(3):
      if board[row, col] == ' ':
        empty_squares.append((row, col))

  return random.choice(empty_squares)


# initialize board
board = np.full((3, 3), ' ', dtype='<U1')

print("Tic-Tac-Toe: Player vs Computer")
choice = input("Do you want to go first or second? (1 or 2): ").strip()

if choice == '1':
  player_mark = 'X'
  computer_mark = 'O'
  player_turn = True
else:
  player_mark = 'O'
  computer_mark = 'X'
  player_turn = False

display_board(board)

while True:
  if player_turn:
    move = input("Enter your move as row,col (0–2): ").strip()

    if ',' not in move:
      print("Invalid format. Use row,col.")
      continue

    parts = move.split(',')
    if not parts[0].isdigit() or not parts[1].isdigit():
      print("Row and column must be integers from 0 to 2.")
      continue

    row = int(parts[0])
    col = int(parts[1])

    if row not in range(3) or col not in range(3):
      print("Row and column must be in the range 0 to 2.")
      continue

    if board[row, col] != ' ':
      print("That square is not empty.")
      continue

    board[row, col] = player_mark
    display_board(board)

    if is_winner(board, player_mark):
      print("You win!")
      break

  else:
    row, col = computer_move(board)
    board[row, col] = computer_mark
    print(f"Computer moves to {row},{col}")
    display_board(board)

    if is_winner(board, computer_mark):
      print("Computer wins!")
      break

  if is_draw(board):
    print("It's a draw!")
    break

  player_turn = not player_turn
```

```python
"""7.12 (Super Challenge Project: 3D Tic-Tac-Toe with Player Against the
Computer) Develop a script that plays three-dimensional Tic-Tac-Toe on a
4-by-4-by-4 board. [Note: This is an extremely challenging project! In the
“Deep Learning” chapter, you’ll learn techniques that will help you
develop and AI-based approach to solving this problem.]"""

"""
Description:
This script plays 3D Tic-Tac-Toe on a 4-by-4-by-4 board between a human
player and the computer. The player chooses whether to go first or second.
Moves are entered as three integers: layer, row, and column (0–3).
The game checks for a winner after each move by evaluating all possible
4-in-a-row winning lines in the 4x4x4 cube. The computer uses a simple
strategy:
1) If it can win in one move, it plays that move.
2) Else, if it must block the player from winning in one move, it blocks.
3) Else, it chooses a random available square.

Input:
- Player choice to go first or second (1 or 2).
- Player moves as "layer,row,col" values from 0 to 3.

Output:
- The 3D board displayed layer by layer after each move.
- Win/draw messages at the end.
"""

import numpy as np
import random


def generate_winning_lines(size=4):
  lines = []

  # rows in each layer
  for z in range(size):
    for y in range(size):
      line = []
      for x in range(size):
        line.append((z, y, x))
      lines.append(line)

  # columns in each layer
  for z in range(size):
    for x in range(size):
      line = []
      for y in range(size):
        line.append((z, y, x))
      lines.append(line)

  # verticals through layers
  for y in range(size):
    for x in range(size):
      line = []
      for z in range(size):
        line.append((z, y, x))
      lines.append(line)

  # diagonals within each layer (2 per layer)
  for z in range(size):
    diag1 = []
    diag2 = []
    for i in range(size):
      diag1.append((z, i, i))
      diag2.append((z, i, size - 1 - i))
    lines.append(diag1)
    lines.append(diag2)

  # diagonals across layers for each row (y fixed)
  for y in range(size):
    diag1 = []
    diag2 = []
    for i in range(size):
      diag1.append((i, y, i))
      diag2.append((i, y, size - 1 - i))
    lines.append(diag1)
    lines.append(diag2)

  # diagonals across layers for each column (x fixed)
  for x in range(size):
    diag1 = []
    diag2 = []
    for i in range(size):
      diag1.append((i, i, x))
      diag2.append((i, size - 1 - i, x))
    lines.append(diag1)
    lines.append(diag2)

  # 4 main space diagonals through the cube
  diag1 = []
  diag2 = []
  diag3 = []
  diag4 = []
  for i in range(size):
    diag1.append((i, i, i))
    diag2.append((i, i, size - 1 - i))
    diag3.append((i, size - 1 - i, i))
    diag4.append((i, size - 1 - i, size - 1 - i))
  lines.append(diag1)
  lines.append(diag2)
  lines.append(diag3)
  lines.append(diag4)

  return lines


def display_board(board):
  print()
  for z in range(4):
    print(f"Layer {z}:")
    for y in range(4):
      print(" ", end="")
      for x in range(4):
        print(board[z, y, x], end="")
        if x < 3:
          print(" | ", end="")
      print()
      if y < 3:
        print("---+---+---+---")
    print()


def is_winner(board, mark, lines):
  for line in lines:
    won = True
    for (z, y, x) in line:
      if board[z, y, x] != mark:
        won = False
        break
    if won:
      return True
  return False


def is_draw(board):
  return not np.any(board == ' ')


def empty_squares(board):
  empties = []
  for z in range(4):
    for y in range(4):
      for x in range(4):
        if board[z, y, x] == ' ':
          empties.append((z, y, x))
  return empties


def find_winning_move(board, mark, lines):
  for (z, y, x) in empty_squares(board):
    board[z, y, x] = mark
    if is_winner(board, mark, lines):
      board[z, y, x] = ' '
      return (z, y, x)
    board[z, y, x] = ' '
  return None


def computer_move(board, computer_mark, player_mark, lines):
  # 1) win if possible
  move = find_winning_move(board, computer_mark, lines)
  if move is not None:
    return move

  # 2) block player if needed
  move = find_winning_move(board, player_mark, lines)
  if move is not None:
    return move

  # 3) otherwise choose a random empty square
  return random.choice(empty_squares(board))


lines = generate_winning_lines(4)
board = np.full((4, 4, 4), ' ', dtype='<U1')

print("3D Tic-Tac-Toe (4x4x4): Player vs Computer")
print("Enter moves as layer,row,col using values 0–3.")
choice = input("Do you want to go first or second? (1 or 2): ").strip()

if choice == '1':
  player_mark = 'X'
  computer_mark = 'O'
  player_turn = True
else:
  player_mark = 'O'
  computer_mark = 'X'
  player_turn = False

display_board(board)

while True:
  if player_turn:
    move = input("Enter your move as layer,row,col (0–3): ").strip()

    if ',' not in move:
      print("Invalid format. Use layer,row,col (example: 0,2,3).")
      continue

    parts = move.split(',')
    if len(parts) != 3:
      print("Invalid format. Use layer,row,col (example: 0,2,3).")
      continue

    if not parts[0].strip().isdigit() or not parts[1].strip().isdigit() or not parts[2].strip().isdigit():
      print("Layer, row, and column must be integers from 0 to 3.")
      continue

    z = int(parts[0].strip())
    y = int(parts[1].strip())
    x = int(parts[2].strip())

    if z not in range(4) or y not in range(4) or x not in range(4):
      print("Layer, row, and column must be in the range 0 to 3.")
      continue

    if board[z, y, x] != ' ':
      print("That square is not empty. Try again.")
      continue

    board[z, y, x] = player_mark
    display_board(board)

    if is_winner(board, player_mark, lines):
      print("You win!")
      break

  else:
    z, y, x = computer_move(board, computer_mark, player_mark, lines)
    board[z, y, x] = computer_mark
    print(f"Computer moves to {z},{y},{x}")
    display_board(board)

    if is_winner(board, computer_mark, lines):
      print("Computer wins!")
      break

  if is_draw(board):
    print("It's a draw!")
    break

  player_turn = not player_turn
```

```python
"""7.13 (Research and Use Other Broadcasting Capabilities) Research the
NumPy broadcasting rules, then create your own arrays to test the rules."""

"""
Numpy Broadcasting rules

When NumPy combines arrays element-by-element (e.g., +, -, *, /),
it compares shapes from the last dimension moving left:
	1.	If the number of dimensions differs, pad the smaller shape with
      1s on the left.  ￼
	2.	Two dimensions are compatible if they are equal, or one of them
      is 1 (the 1 “stretches”).  ￼
	3.	If any dimension pair is not equal and neither is 1, broadcasting
      fails with a ValueError.  ￼
"""

"""
Description:
Tests NumPy broadcasting rules by creating arrays with different shapes,
performing operations, and printing the resulting shapes/values.
"""

import numpy as np

# 1) Scalar broadcasts to any shape
a = np.array([1, 2, 3])
print("1) scalar + vector")
print("a.shape:", a.shape)
print("(a + 10):", a + 10)
print()

# 2) (3, 1) + (1, 4) -> (3, 4)  (classic broadcast grid)
b = np.array([[1], [2], [3]])        # shape (3, 1)
c = np.array([[10, 20, 30, 40]])     # shape (1, 4)
print("2) (3,1) + (1,4) -> (3,4)")
print("b.shape:", b.shape, "c.shape:", c.shape)
print("result:\n", b + c)
print("result.shape:", (b + c).shape)
print()

# 3) (2, 3) + (3,) -> (2, 3)  (1D treated like (1,3))
d = np.array([[1, 2, 3],
              [4, 5, 6]])            # shape (2, 3)
e = np.array([10, 20, 30])           # shape (3,)
print("3) (2,3) + (3,) -> (2,3)")
print("d.shape:", d.shape, "e.shape:", e.shape)
print("result:\n", d + e)
print()

# 4) Broadcasting failure example: (2, 3) and (2, 2) are incompatible
f = np.ones((2, 3), dtype=int)
g = np.ones((2, 2), dtype=int)
print("4) incompatible shapes (2,3) and (2,2)")
print("f.shape:", f.shape, "g.shape:", g.shape)
try:
  print(f + g)
except ValueError as error:
  print("Broadcasting error:", error)
print()

# 5) 3D + 2D: (2,3,4) + (3,4) -> (2,3,4)  (2D treated like (1,3,4))
h = np.arange(24).reshape(2, 3, 4)
i = np.ones((3, 4), dtype=int)
print("5) (2,3,4) + (3,4) -> (2,3,4)")
print("h.shape:", h.shape, "i.shape:", i.shape)
result = h + i
print("result.shape:", result.shape)
print("result[0]:\n", result[0])
```

```python
"""7.18 (Median and Mode of an array) NumPy arrays offer a mean
method, but not median or mode. Write functions median and mode
that use existing NumPy capabilities to determine the median (middle) and
mode (most frequent) of the values in an array. Your functions should
determine the median and mode regardless of the array’s shape. Test your
function on three arrays of different shapes."""

import numpy as np


def median(array):
  """
  Returns the median value of the array, regardless of shape.
  """
  flattened = array.flatten()
  sorted_values = np.sort(flattened)
  n = len(sorted_values)

  if n % 2 == 1:
    return sorted_values[n // 2]
  else:
    return (sorted_values[n // 2 - 1] + sorted_values[n // 2]) / 2


def mode(array):
  """
  Returns the mode (most frequent value) of the array.
  """
  flattened = array.flatten()
  counts = np.bincount(flattened)
  return np.argmax(counts)

# Testing on arrays of different shapes

array1 = np.array([1, 2, 2, 3, 4, 5, 2])
array2 = np.array([[5, 1, 3],
                   [2, 2, 4]])
array3 = np.array([[[7, 8], [8, 9]],
                   [[8, 10], [11, 8]]])

arrays = [array1, array2, array3]

for i, arr in enumerate(arrays, start=1):
  print(f"Array {i}:")
  print(arr)
  print("Median:", median(arr))
  print("Mode:", mode(arr))
  print()
```
