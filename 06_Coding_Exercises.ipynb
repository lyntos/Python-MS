11.1 What does the following code do?
```
In [1]: def mystery(a, b):
...: if b == 1:
...: return a
...: else:
...: return a + mystery(a, b - 1)
...:
In [2]: mystery(2, 10)
Out[2]: ?????
```

Answer: The code above is a recursive function. It will repeatedly add `a` to itself "b times".

So `mystery(2,10)` will keep adding a to itself for 10 times until it meet the condition of b being equal to 1 - `mystery(2, 1)`. This will result in an output of `20`

````python
def mystery(a, b):
  if b == 1:
    return a
  else:
    print(a + mystery(a, b - 1))
    return a + mystery(a, b - 1)

mystery(2, 10)
````

11.2 Find the logic error(s) in the following recursive function, and
explain how to correct it (them). This function should find the sum of the
values from 0 to n.
```
In [3]: def sum(n):
...: if n == 0:
...: return 0
...: else:
...: return n + sum(n)
...:
```

Answer: The function `sum(n)` fails one of the basic rules of recursion - the recursive call has to change or reduce the value of the previous argument passed into the function call again.

`return n + sum(n)` does not do this, so the function will end in an infinite error (RecursionError).
The block below shows an updated code

````python
"""Fixed code"""
def sum(n):
  if n == 0:
    return 0
  else:
    return n + sum(n - 1)
````

11.5 (Recursive power Function) Write a recursive function
power(base, exponent) that, when called, returns
```
base ^exponent
```
For example, power(3,4) = 3 * 3 * 3 * 3. Assume that exponent is
an integer greater than or equal to 1. Hint: The recursion step should use
the relationship
```
base^e = base · base^e-1
```
and the terminating condition occurs when exponent is equal to 1,
because
```
base^1 = base
```
Incorporate this function into a program that enables the user to enter the
base and exponent.

````python
def power(base, exponent):
  """
  Description: Takes a base number and an exponent and returns the base
  to the power of the exponent.
  """
  if exponent == 1:
    return base
  else:
    return base * power(base, exponent - 1)

print(power(3, 4))
````

11.6 (Recursive Fibonacci Modification) Modify Section 11.4’s recursive
fibonacci function to keep track of the total number of function calls.
Display the number of calls for fibonacci(10), fibonacci(20)
and fibonacci(30).

````python
call_count = 0

def fibonacci(n):
  global call_count
  call_count += 1

  if n <= 1:
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)


# Test values
for n in [10, 20, 30]:
  call_count = 0
  result = fibonacci(n)
  print(f"fibonacci({n}) = {result}")
  print(f"Total function calls: {call_count}\n")
````

11.7 (Improving Recursive Fibonacci Performance: Memoization)
Research the performance enhancement technique called memoization.
Modify Section 11.4’s recursive fibonacci function to incorporate
memoization. Compare the performance of both versions for
`fibonacci(10)`, `fibonacci(20)` and `fibonacci(30)`.

````python
call_count = 0

def fibonacci_memo(n, memo={}):
  global call_count
  call_count += 1

  if n in memo:
    return memo[n]

  if n <= 1:
    memo[n] = n
  else:
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)

  return memo[n]


# Compare versions
for n in [10, 20, 30]:
  call_count = 0
  result = fibonacci_memo(n, {})
  print(f"Memoized fibonacci({n}) = {result}")
  print(f"Total function calls: {call_count}\n")
````

11.8 (Visualizing Recursion) It’s interesting to watch recursion “in action.”
Modify the recursive factorial function presented in this chapter to
print its local variable and recursive-call parameter. For each recursive
call, display the outputs on a separate line and add a level of indentation.
Make the outputs clear, interesting and meaningful. Your goal here is to
design and implement an output format that makes it easier to understand
recursion.

````python
def factorial(n, depth=0):
  indent = "  " * depth

  print(f"{indent}Entering factorial({n})")

  if n == 1:
    print(f"{indent}Base case reached: factorial(1) = 1")
    return 1
  else:
    print(f"{indent}Computing {n} * factorial({n-1})")

    result = n * factorial(n - 1, depth + 1)

    print(f"{indent}Returning {result} from factorial({n})")
    return result


# Test
print("\nFinal Result:", factorial(5))
````

11.9 (Greatest Common Divisor) The greatest common divisor of integers
x and y is the largest integer that evenly divides into both x and y. Write
and test a recursive function gcd that returns the greatest common divisor
of x and y. The gcd of x and y is defined recursively as follows: If y is
equal to 0, then gcd(x, y) is x; otherwise, gcd(x, y) is gcd(y, x %
y), where % is the remainder operator.

````python
"""
  Description: Returns the greatest common divisor of the given integers
"""
def gcd(x, y):
  if y == 0:
    return x
  else:
    return gcd(y, x % y)

print(gcd(48, 18))
````

11.10 (Palindromes) A palindrome is a string that is spelled the same way
forward and backward. Some examples of palindromes are “radar,” “able
was i ere i saw elba” and (if spaces are ignored) “a man a plan a canal
panama.” Write a recursive test_palindrome function that returns
True if the string stored in an array is a palindrome and False
otherwise. The function should ignore spaces and punctuation in the
string.

````python
import re

def test_palindrome(text):
  """
  Description: Returns True if the given string is a palindrome and False otherwise.
  """
  # Remove spaces, punctuation, make lowercase
  cleaned = re.sub(r'[^a-zA-Z0-9]', '', text).lower()

  def recursive_check(s):
    # Base case: length 0 or 1
    if len(s) <= 1:
      return True

    # If first and last characters differ = not palindrome
    if s[0] != s[-1]:
      return False

    # Recursive step: check inner substring
    return recursive_check(s[1:-1])

  return recursive_check(cleaned)

print(test_palindrome("radar"))
print(test_palindrome("Able was I ere I saw Elba"))
print(test_palindrome("A man, a plan, a canal, Panama!"))
print(test_palindrome("hello"))
````

11.18 (Recursive Binary Search) Modify this chapter’s binary_search
function to perform a recursive binary search of the array. The function
should receive the search key, starting index and ending index as
arguments. If the search key is found, return its index in the array. If the
search key is not found, return –1.

````python
def binary_search(array, key, low, high):
  # Base case: not found
  if low > high:
    return -1

  mid = (low + high) // 2

  if array[mid] == key:
    return mid
  elif key < array[mid]:
    return binary_search(array, key, low, mid - 1)
  else:
    return binary_search(array, key, mid + 1, high)

numbers = [2, 5, 8, 12, 16, 23, 38, 56, 72]

print(binary_search(numbers, 23, 0, len(numbers) - 1))  # Found
print(binary_search(numbers, 10, 0, len(numbers) - 1))  # Not found
````
