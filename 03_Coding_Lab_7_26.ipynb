```python
"""7.26 (Knight’s Tour Project: Closed-Tour Test) In the Knight’s Tour, a full
tour occurs when the knight makes 64 moves, touching each square of the
chessboard once and only once. A closed tour occurs when the 64th move
is one move away from the square in which the knight started the tour.
Modify the script you wrote in Exercise 7.24 to test for a closed tour if a
full tour has occurred."""

"""
Description:
This script performs a Knight's Tour on an 8-by-8 chessboard using the
accessibility heuristic (Warnsdorff-style). If a full tour (64 moves) occurs,
the script also tests whether the tour is a closed tour—meaning the final
position is one knight move away from the starting position.
"""

import numpy as np
import random

BOARD_SIZE = 8

horizontal = np.array([2, 1, -1, -2, -2, -1, 1, 2])
vertical = np.array([-1, -2, -2, -1, 1, 2, 2, 1])


def in_bounds(r, c):
  return 0 <= r < BOARD_SIZE and 0 <= c < BOARD_SIZE


def initial_accessibility():
  return np.array([
    [2, 3, 4, 4, 4, 4, 3, 2],
    [3, 4, 6, 6, 6, 6, 4, 3],
    [4, 6, 8, 8, 8, 8, 6, 4],
    [4, 6, 8, 8, 8, 8, 6, 4],
    [4, 6, 8, 8, 8, 8, 6, 4],
    [4, 6, 8, 8, 8, 8, 6, 4],
    [3, 4, 6, 6, 6, 6, 4, 3],
    [2, 3, 4, 4, 4, 4, 3, 2]
  ], dtype=int)


def update_accessibility(access, row, col, board):
  for move in range(8):
    r = row + vertical[move]
    c = col + horizontal[move]
    if in_bounds(r, c) and board[r, c] == 0:
      access[r, c] -= 1


def is_knight_move_away(r1, c1, r2, c2):
  for move in range(8):
    if r1 + vertical[move] == r2 and c1 + horizontal[move] == c2:
      return True
  return False


def knights_tour_heuristic(start_row, start_col):
  board = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=int)
  access = initial_accessibility()

  current_row = start_row
  current_col = start_col
  move_number = 1
  board[current_row, current_col] = move_number
  access[current_row, current_col] = 99

  update_accessibility(access, current_row, current_col, board)

  while move_number < 64:
    best_moves = []
    best_value = 10**9

    for move in range(8):
      next_row = current_row + vertical[move]
      next_col = current_col + horizontal[move]

      if in_bounds(next_row, next_col) and board[next_row, next_col] == 0:
        value = access[next_row, next_col]
        if value < best_value:
          best_value = value
          best_moves = [(next_row, next_col)]
        elif value == best_value:
          best_moves.append((next_row, next_col))

    if not best_moves:
      break

    current_row, current_col = random.choice(best_moves)
    move_number += 1
    board[current_row, current_col] = move_number
    access[current_row, current_col] = 99
    update_accessibility(access, current_row, current_col, board)

  full_tour = (move_number == 64)

  if full_tour:
    closed_tour = is_knight_move_away(current_row, current_col, start_row, start_col)
  else:
    closed_tour = False

  return board, move_number, full_tour, closed_tour


def display_board(board):
  width = len(str(board.max()))
  for r in range(BOARD_SIZE):
    for c in range(BOARD_SIZE):
      print(f"{board[r, c]:>{width}}", end=" ")
    print()
  print()

# Demo run - start row and column can be changed when playing

start_row = 5
start_col = 5

board, moves, full, closed = knights_tour_heuristic(start_row, start_col)

display_board(board)
print("Moves completed:", moves)

if full:
  print("Full tour: YES")
  print("Closed tour:", "YES" if closed else "NO")
else:
  print("Full tour: NO")
  print("Closed tour: NO")
```
