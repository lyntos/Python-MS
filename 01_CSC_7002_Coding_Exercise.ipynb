```python
"""
2.12 (7% Investment Return) Some investment advisors say that it’s
reasonable to expect a 7% return over the long term in the stock market.
Assuming that you begin with $1000 and leave your money invested,
calculate and display how much money you’ll have after 10, 20 and 30
years. Use the following formula for determining these amounts:
n
a = p(1 + r)^2
where
p is the original amount invested (i.e., the principal of $1000),
r is the annual rate of return (7%),
n is the number of years (10, 20 or 30) and
a is the amount on deposit at the end of the
nth year.
"""
def calculate_investment_growth():
  """
  Description:
  This function calculates the future value of an investment based on a fixed
  annual rate of return. It assumes an initial investment of $1000 and a 7%
  annual return. Using the compound interest formula a = p(1 + r)^n, the
  function computes and displays the amount of money on deposit after
  10, 20, and 30 years.

  Input:
  - No user input is required.
  - The principal amount (p = 1000), annual rate of return (r = 0.07),
    and investment durations (10, 20, and 30 years) are predefined
    within the function.

  Output:
  - The amount of money on deposit after 10 years.
  - The amount of money on deposit after 20 years.
  - The amount of money on deposit after 30 years.
  - All results are displayed in dollar format with two decimal places.
  """

  principal = 1000 # Original investment amount (p)
  rate = 0.07 # Annual rate of return (r)

  years_10 = 10
  years_20 = 20
  years_30 = 30

  amount_10 = principal * (1 + rate) ** years_10
  amount_20 = principal * (1 + rate) ** years_20
  amount_30 = principal * (1 + rate) ** years_30

  print("Investment growth at 7% annual return")
  print("-------------------------------------")
  print(f"After 10 years: ${amount_10:.2f}")
  print(f"After 20 years: ${amount_20:.2f}")
  print(f"After 30 years: ${amount_30:.2f}")


calculate_investment_growth()
```

```python
""" 3.1 (Validating User Input) Modify the script of Fig3.3. to validate its
inputs. For any input, if the value entered is other than 1 or 2, keep looping
until the user enters a correct value. Use one counter to keep track of the
number of passes, then calculate the number of failures after all the user’s
inputs have been received."""

"""
Description:
This program processes examination results for 10 students.
Each student enters a result of 1 (pass) or 2 (fail).
If an invalid value is entered, the program repeatedly prompts
the user until a valid value is provided. The program counts
the number of passes and calculates the number of failures
after all inputs have been received.

Input:
- Ten integer values entered by the user.
- Each value must be either 1 (pass) or 2 (fail).

Output:
- Total number of students who passed.
- Total number of students who failed.
- A bonus message if more than 8 students passed.

Example:
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 2
Enter result (1=pass, 2=fail): 3
Invalid input. Please enter 1 or 2.
Enter result (1=pass, 2=fail): 3
Invalid input. Please enter 1 or 2.
Enter result (1=pass, 2=fail): 2
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 2
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 1
Enter result (1=pass, 2=fail): 1
Passed: 7
Failed: 3
"""

# initialize variables
passes = 0          # number of passes
student = 1         # student counter

# process 10 students
while student <= 10:

  # get exam result
  result = int(input("Enter result (1=pass, 2=fail): "))

  # validate input
  while result != 1 and result != 2:
      print("Invalid input. Please enter 1 or 2.")
      result = int(input("Enter result (1=pass, 2=fail): "))

  if result == 1:
      passes = passes + 1

  student = student + 1

# termination phase
failures = 10 - passes

print("Passed:", passes)
print("Failed:", failures)

if passes > 8:
  print("Bonus to instructor")
```

```python
"""3.10 (7% Investment Return) Reimplement Exercise 2.12 to use a loop
that calculates and displays the amount of money you’ll have each year at
the ends of years 1 through 30."""
"""
Description:
This program calculates and displays the future value of an investment
for each year from 1 through 30. The program assumes an initial investment
of $1000 and a fixed annual rate of return of 7%. A loop is used to compute
and display the amount of money on deposit at the end of each year using
the compound interest formula a = p(1 + r)^n.

Input:
- No user input is required.
- The principal amount and annual rate of return are predefined.

Output:
- The amount of money on deposit at the end of each year from year 1
  through year 30, formatted to two decimal places.

Sample output:
Year   Amount on Deposit
--------------------------
   1   $1,070.00
   2   $1,144.90
   3   $1,225.04
   ...
  30   $7,612.26
"""

principal = 1000
rate = 0.07

print(f"{'Year':>6}   {'Amount on Deposit'}")
print("--------------------------")

for year in range(1, 31):
  amount = principal * (1 + rate) ** year
  print(f"{year:>4}   ${amount:,.2f}")
```

```python
"""3.11 (Miles Per Gallon) Drivers are concerned with the mileage obtained
by their automobiles. One driver has kept track of several tankfuls of
gasoline by recording miles driven and gallons used for each tankful.
Develop a sentinel-controlled-repetition script that prompts the user to
input the miles driven and gallons used for each tankful. The script should
calculate and display the miles per gallon obtained for each tankful. After
processing all input information, the script should calculate and display
the combined miles per gallon obtained for all tankfuls (that is, total miles
driven divided by total gallons used).
Enter the gallons used (-1 to end): 12.8
Enter the miles driven: 287
The miles/gallon for this tank was 22.421875
Enter the gallons used (-1 to end): 10.3
Enter the miles driven: 200
The miles/gallon for this tank was 19.417475
Enter the gallons used (-1 to end): 5
Enter the miles driven: 120
The miles/gallon for this tank was 24.000000
Enter the gallons used (-1 to end): -1
The overall average miles/gallon was 21.601423
"""

"""
Description:
This program calculates the miles per gallon (MPG) for each tankful of
gasoline entered by the user. The user repeatedly enters the gallons
used and the miles driven for each tank. The program displays the MPG
for each tankful and, after all input is complete, displays the overall
average MPG for all tankfuls combined.

Input:
- Gallons used for each tankful (a value of -1 ends input).
- Miles driven for each tankful.

Output:
- Miles per gallon for each tankful.
- Overall average miles per gallon for all tankfuls.
"""

total_miles = 0.0
total_gallons = 0.0

gallons = float(input("Enter the gallons used (-1 to end): "))

while gallons != -1:
  miles = float(input("Enter the miles driven: "))

  mpg = miles / gallons
  print(f"The miles/gallon for this tank was {mpg:.6f}\n")

  total_miles = total_miles + miles
  total_gallons = total_gallons + gallons

  gallons = float(input("Enter the gallons used (-1 to end): "))

if total_gallons != 0:
  overall_mpg = total_miles / total_gallons
  print(f"The overall average miles/gallon was {overall_mpg:.6f}")
```

```python
"""3.17 (Nested Loops) Write a script that displays the following triangle
patterns separately, one below the other. Separate each pattern from the
next by one blank line. Use for loops to generate the patterns. Display all
asterisks ( *) with a single statement of the form

print('*', end='')

which causes the asterisks to display side by side. [Hint: For the last two
patterns, begin each line with zero or more space characters.]
"""

"""
Description:
This program displays four different triangle patterns made of asterisks.
Each pattern is displayed separately using nested for loops. A single
statement of the form print('*', end='') is used to display the asterisks.
Each pattern is separated from the next by one blank line.

Input:
- No user input is required.

Output:
- Four triangle patterns displayed one below the other, separated by
  a blank line.
"""

# (a) Left-aligned increasing triangle
for row in range(1, 11):
  for column in range(row):
    print('*', end='')
  print()

print()

# (b) Left-aligned decreasing triangle
for row in range(10, 0, -1):
  for column in range(row):
    print('*', end='')
  print()

print()

# (c) Right-aligned decreasing triangle
for row in range(10, 0, -1):
  for space in range(10 - row):
    print(' ', end='')
  for column in range(row):
    print('*', end='')
  print()

print()

# (d) Right-aligned increasing triangle
for row in range(1, 11):
  for space in range(10 - row):
    print(' ', end='')
  for column in range(row):
    print('*', end='')
  print()
```

```python
"""3.19 (Brute-Force Computing: Pythagorean Triples) A right triangle can
have sides that are all integers. The set of three integer values for the sides
of a right triangle is called a Pythagorean triple. These three sides must
satisfy the relationship that the sum of the squares of two of the sides is
equal to the square of the hypotenuse. Find all Pythagorean triples for
side1, side2 and hypotenuse (such as 3, 4 and 5) all no larger than
20. Use a triple-nested for-loop that tries all possibilities. This is an
example of “brute-force” computing. You’ll learn in more advanced
computer science courses that there are many interesting problems for
which there is no known algorithmic approach other than sheer brute
force."""

"""
Description:
This program finds all Pythagorean triples with integer sides no larger
than 20. A Pythagorean triple consists of three integers such that the
sum of the squares of two sides is equal to the square of the
hypotenuse. The program uses a brute-force approach with a triple-nested
for loop to test all possible combinations.

Input:
- No user input is required.

Output:
- All Pythagorean triples (side1, side2, hypotenuse) where each value
  is less than or equal to 20.

Example output
3 4 5
4 3 5
5 12 13
12 5 13
8 15 17
15 8 17
12 16 20
16 12 20
"""

print("Pythagorean triples (side1, side2, hypotenuse):")

for side1 in range(1, 21):
  for side2 in range(1, 21):
    for hypotenuse in range(1, 21):
      if side1 ** 2 + side2 ** 2 == hypotenuse ** 2:
        print(side1, side2, hypotenuse)
```

```python
"""3.20 (Binary-to-Decimal Conversion) Input an integer containing 0s and
1s (i.e., a “binary” integer) and display its decimal equivalent. The online
appendix, “Number Systems,” discusses the binary number system. [Hint:
Use the modulus and division operators to pick off the “binary” number’s
digits one at a time from right to left. Just as in the decimal number
system, where the rightmost digit has the positional value 1 and the next
digit to the left has the positional value 10, then 100, then 1000, etc., in the
binary number system, the rightmost digit has the positional value 1, the
next digit to the left has the positional value 2, then 4, then 8, etc. Thus,
the decimal number 234 can be interpreted as 2 * 100 + 3 * 10 + 4 * 1.
The decimal equivalent of binary 1101 is 1 * 8 + 1 * 4 + 0 * 2 + 1 * 1.]"""

"""
Description:
This program converts a binary number (containing only digits 0 and 1)
into its decimal equivalent. The program processes the binary number
from right to left using the modulus (%) and integer division (//)
operators. Each binary digit is multiplied by its positional value,
which doubles each iteration.

Input:
- An integer containing only the digits 0 and 1.

Output:
- The decimal equivalent of the binary number.

Example:
Input - 1101
Output - 13
"""

binary = int(input("Enter a binary number: "))

decimal = 0
power = 1

while binary > 0:
  digit = binary % 10
  decimal = decimal + digit * power
  binary = binary // 10
  power = power * 2

print("The decimal equivalent is:", decimal)
```
