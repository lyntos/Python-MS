```python
"""8.1 (Check Protection) Although electronic deposit has become extremely
popular, payroll and accounts payable applications often print checks. A
serious problem is the intentional alteration of a check amount by
someone who plans to cash a check fraudulently. To prevent a dollar
amount from being altered, some computerized check-writing systems
employ a technique called check protection. Checks designed for printing
by computer typically contain a fixed number of spaces for the printed
amount. Suppose a paycheck contains eight blank spaces in which the
computer is supposed to print the amount of a weekly paycheck. If the
amount is large, then all eight of the spaces will be filled:
1,230.60 (check amount)
--------
01234567 (position numbers)
On the other hand, if the amount is smaller, then several of the spaces
would ordinarily be left blank. For example,
399.87
--------
01234567
contains two blank spaces. If a check is printed with blank spaces, it’s
easier for someone to alter the amount. Check-writing systems often insert
leading asterisks to prevent alteration and protect the amount as follows:
**399.87
--------
01234567
Write a script that inputs a dollar amount, then prints the amount in
checkprotected format in a field of 10 characters with leading asterisks if
necessary. [Hint: In a format string that explicitly specifies alignment with
<, ^ or >, you can precede the alignment specifier with the fill character of
your choice.]"""

amount = float(input("Enter the dollar amount: \n"))

# Format:
# * --> fill character
# > --> right align
# 10 --> total width
# .2f --> 2 decimal places
protected_amount = f"{amount:*>10.2f}"

print("Check-protected amount:")
print(protected_amount)
```

```python
"""8.2 (Random Sentences) Write a script that uses random-number
generation to compose sentences. Use four arrays of strings called
article, noun, verb and preposition. Create a sentence by
selecting a word at random from each array in the following order:
article, noun, verb, preposition, article and noun. As each
word is picked, concatenate it to the previous words in the sentence.
Spaces should separate the words. When the final sentence is output, it
should start with a capital letter and end with a period. The script should
generate and display 20 sentences."""

"""
Description:
This script generates 20 random sentences using four arrays:
article, noun, verb, and preposition.
Each sentence follows the pattern:
"article noun verb preposition article noun".
"""
import random

# Word arrays
article = ["the", "a", "one", "some", "any"]
noun = ["boy", "girl", "dog", "town", "car"]
verb = ["drove", "jumped", "ran", "walked", "skipped"]
preposition = ["to", "from", "over", "under", "on"]

# Generate 20 sentences
for _ in range(20):
  sentence = (
    random.choice(article) + " " +
    random.choice(noun) + " " +
    random.choice(verb) + " " +
    random.choice(preposition) + " " +
    random.choice(article) + " " +
    random.choice(noun)
  )

  # Capitalize first letter and add period
  sentence = sentence.capitalize() + "."

  print(sentence)
```

```python
"""8.3 (Pig Latin) Write a script that encodes English-language phrases into a
form of coded language called pig Latin. There are many different ways to
form pig Latin phrases. For simplicity, use the following algorithm:
To form a pig Latin phrase from an English-language phrase, tokenize the
phrase into words with string method split. To translate each English
word into a pig Latin word, place the first letter of the English word at the
end of the word and add the letters “ay.” Thus, the word “jump” becomes
“umpjay,” the word “the” becomes “hetay,” and the word “computer”
becomes “omputercay.” If the word starts with a vowel, just add “ay.”
Blanks between words remain as blanks. Assume the following: The
English phrase consists of words separated by blanks, there are no
punctuation marks and all words have two or more letters. Enable the user
to enter a sentence, then display the sentence in pig Latin."""

"""
Description:
This script encodes an English-language phrase into Pig Latin using
a simple algorithm:
- Move the first letter of each word to the end of the word.
- Append 'ay' to the end.

Input:
- A phrase (string) entered by the user.

Output:
- The phrase converted to Pig Latin.
"""

phrase = input("Enter an English phrase: ").strip()
words = phrase.split()

pig_latin_words = []

for word in words:
  if len(word) > 0:
    pig_word = word[1:] + word[0] + "ay"
    pig_latin_words.append(pig_word)

pig_latin_phrase = " ".join(pig_latin_words)
print("Pig Latin:", pig_latin_phrase)
```

```python
"""8.7 (Converting Integers to Characters) Use the c presentation type to
display a table of the character codes in the range 0 to 255 and their
corresponding characters."""
"""
Description:
This script displays a table of character codes from 0 to 255
and their corresponding characters using the 'c' presentation type.
"""

print(f"{'Code':>5} {'Character':>10}")
print("-" * 20)

for code in range(256):
  try:
    print(f"{code:5} {code:10c}")
  except ValueError:
    # Some control characters may not display nicely
    print(f"{code:5} {' ':>10}")
```

```python
"""8.10 (Project: Simple Sentiment Analysis) Search online for lists of
positive sentiment words and negative sentiment words. Create a script
that inputs text, then determines whether that text is positive or negative,
based on the total number of positive words and the total number of
negative words. Test your script by searching for Twitter tweets on a topic
of your choosing, then entering the text for several tweets. In the data
science case study chapters, we’ll take a deeper look at sentiment analysis."""

"""
Description:
This script performs simple sentiment analysis using a positive-word list
and a negative-word list. It reads text from the user, counts how many
positive and negative words appear, then labels the text as Positive,
Negative, or Neutral.

Input:
- A line of text entered by the user (e.g., a tweet).

Output:
- Number of positive words found
- Number of negative words found
- Overall sentiment label

Test with -
Negative words - https://github.com/shekhargulati/sentiment-analysis-python/blob/master/opinion-lexicon-English/negative-words.txt
Positive words - https://github.com/shekhargulati/sentiment-analysis-python/blob/master/opinion-lexicon-English/positive-words.txt
"""


import re


def load_word_list(filename):
  # Loads words from a file, ignoring blank lines and comment lines.
  words = set()

  with open(filename, "r", encoding="utf-8", errors="ignore") as file:
    for line in file:
      line = line.strip()
      if line == "" or line.startswith(";"):
        continue
      words.add(line.lower())

  return words


positive_words = load_word_list("positive-words.txt")
negative_words = load_word_list("negative-words.txt")

print("Enter text to analyze (press Enter with no text to stop).")

while True:
  text = input("\nText: ").strip()
  if text == "":
    break

  # tokenize: keep letters, numbers, and apostrophes (simple approach)
  tokens = re.findall(r"[a-zA-Z0-9']+", text.lower())

  positive_count = 0
  negative_count = 0

  for token in tokens:
    if token in positive_words:
      positive_count += 1
    if token in negative_words:
      negative_count += 1

  if positive_count > negative_count:
    sentiment = "Positive"
  elif negative_count > positive_count:
    sentiment = "Negative"
  else:
    sentiment = "Neutral"

  print("Positive words:", positive_count)
  print("Negative words:", negative_count)
  print("Sentiment:", sentiment)

  """
  This works, but found out a phrase like
  'Oh get away from me. You do not make me happy' returns a Positive sentiment,
  which is False. This is something I can optimize in the future, but that will
  mean I have to also consider the words before and after any word in the
  matching list.

  Should be a fun addition :)
  """
```

```python
"""8.11 (Project: Evaluate Word Problems) Write a script that enables the user
to enter mathematical word problems like “two times three” and “seven
minus five”, then use string processing to break apart the string into the
numbers and the operation and return the result. So “two times three”
would return 6 and “seven minus five” would return 2. To keep things
simple, assume the user enters only the words for the numbers 0 through 9
and only the operations 'plus', 'minus', 'times' and 'divided
by'."""

"""
Description:
This script evaluates simple mathematical word problems such as
"two times three" or "seven minus five".
"""

# Mapping word numbers to integers
number_words = {
  "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
  "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9
}

while True:
  expression = input("Enter a math expression (or press Enter to quit): ").strip().lower()
  if expression == "":
    break

  tokens = expression.split()

  # Handle "divided by" specially
  if "divided" in tokens:
    num1 = number_words[tokens[0]]
    num2 = number_words[tokens[-1]]
    result = num1 / num2
  else:
    num1 = number_words[tokens[0]]
    operation = tokens[1]
    num2 = number_words[tokens[2]]

    if operation == "plus":
      result = num1 + num2
    elif operation == "minus":
      result = num1 - num2
    elif operation == "times":
      result = num1 * num2

  print("Result:", result)
```

```python
"""8.12 (Project: Scrambled Text) Use string-processing capabilities to keep
the first and last letter of a word and scramble the remaining letters in
between the first and last. Search online for “University of Cambridge
scrambled text” for an intriguing paper on the readability of texts
consisting of such scrambled words. Investigate the random module’s
shuffle function to help you implement this exercise’s solution."""

"""
Description:
This script scrambles the internal letters of each word while
keeping the first and last letters unchanged.
"""

import random

def scramble_word(word):
  if len(word) <= 3:
      return word

  first = word[0]
  middle = list(word[1:-1])
  last = word[-1]

  random.shuffle(middle)

  return first + "".join(middle) + last

sentence = input("Enter a sentence: ")

words = sentence.split()

scrambled = [scramble_word(word) for word in words]

print("Scrambled text:")
print(" ".join(scrambled))
```

```python
"""8.13 (Regular Expressions: Condense Spaces to a Single Space) Check
whether a sentence contains more than one space between words. If so,
remove the extra spaces and display the results. For example, 'Hello
World' should become 'Hello World'."""

"""
Description:
This script checks whether a sentence contains more than one space between
words. If so, it removes extra spaces and displays the result.
"""

import re

sentence = input("Enter a sentence: ")
count = 0

if re.search(r"\s{2,}", sentence):
  count = len(re.findall(r"\s{2,}", sentence)) # extra
  fixed = re.sub(r"\s{2,}", " ", sentence)
  print("Condensed:", fixed)
  print("Number of spaces more than 1: ", count)
else:
  print("No extra spaces found.")
  print("Same:", sentence)
```

```python
"""8.15 (Regular Expressions: Counting Characters and Words) Use regular
expressions and the findall function to count the number of digits,
non-digit characters, whitespace characters and words in a string."""

"""
Description:
This script uses regular expressions and findall to count the number of:
digits, non-digit characters, whitespace characters, and words in a string.
"""

import re

text = input("Enter a string: ")

digits = re.findall(r"\d", text)
non_digits = re.findall(r"\D", text)
whitespace = re.findall(r"\s", text)
words = re.findall(r"\b\w+\b", text)

print("Digits:", len(digits))
print("Non-digit characters:", len(non_digits))
print("Whitespace characters:", len(whitespace))
print("Words:", len(words))
```

```python
"""8.16 (Regular Expressions: Locating URLs) Use a regular expression to
search through a string and to locate all valid URLs. For this exercise,
assume that a valid URL has the form
http://www.domain_name.extension, where extension must be two or
more characters."""

"""
Description:
This script uses a regular expression to locate all valid URLs in a string.
A valid URL has the form:
http://www.domain_name.extension
where extension is at least 2 characters.
"""

import re

text = input("Enter text containing URLs: ")

pattern = r"http://www\.[A-Za-z0-9_]+\.[A-Za-z]{2,}"

matches = re.findall(pattern, text)

print("URLs found:")
for url in matches:
  print(url)
```

```python
"""8.17 (Regular Expressions: Matching Numeric Values) Write a regular
expression that searches a string and matches a valid number. A number
can have any number of digits, but it can have only digits and a decimal
point and possibly a leading sign. The decimal point is optional, but if it
appears in the number, there must be only one, and it must have digits on
its left and its right. There should be whitespace or a beginning or end-ofline
character on either side of a valid number."""

"""
Description:
This script finds valid numbers in a string using a regular expression.
A valid number may have a leading sign and an optional decimal point,
but if a decimal point is present it must have digits on both sides.
Valid numbers must be separated by whitespace or line boundaries.
"""

import re

text = input("Enter a string containing numbers: ")

pattern = r'(^|\s)[+-]?\d+(\.\d+)?(?=\s|$)'

matches = re.finditer(pattern, text)

print("Valid numbers found:")

for match in matches:
  print(match.group().strip())
```

```python
"""8.18 (Regular Expression: Password Format Validator) Search online for
secure password recommendations, then research existing regular
expressions that validate secure passwords. Two examples of password
requirements are:
1. Passwords must contain at least five words, each separated by a
hyphen, a space, a period, a comma or an underscore.
2. Passwords must have a minimum of 8 characters and contain at
least one each from uppercase characters, lowercase characters,
digits and punctuation characters (such as characters in '!@#$%
<^>&*?').
Write regular expressions for each of the two requirements above, then use
them to test sample passwords."""

"""
Description:
This script validates passwords against two different regex requirements:
1) At least five words separated by - space . , or _
2) At least 8 characters with uppercase, lowercase, digit, and punctuation
"""

import re

pattern1 = r'^[A-Za-z]+([\- \.,_][A-Za-z]+){4,}$'
pattern2 = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[!@#$%\^&*?]).{8,}$'

passwords = [
  "one-two-three-four-five",
  "alpha beta gamma delta epsilon",
  "a.b.c.d.e",
  "one_two_three_four",              # invalid (only 4 words)
  "Abc123!@",                        # valid for requirement 2
  "abc123!@",                        # invalid (no uppercase)
  "ABC123!@",                        # invalid (no lowercase)
  "Abcdefgh",                        # invalid (no digit, no punctuation)
  "Abc12345"                         # invalid (no punctuation)
]

print("Requirement 1 (5+ words separated by - space . , _):")
for pw in passwords:
  print(pw, "->", "Valid" if re.fullmatch(pattern1, pw) else "Invalid")

print("\nRequirement 2 (8+ chars, upper, lower, digit, punctuation):")
for pw in passwords:
  print(pw, "->", "Valid" if re.fullmatch(pattern2, pw) else "Invalid")
```

```python
"""8.19 (Regular Expressions: Testing Regular Expressions Online) Before
using any regular expression in your code, you should thoroughly test it to
ensure that it meets your needs. Use a regular expression website like
regex101.com to explore and test existing regular expressions, then
write your own regular expression tester."""

"""
Description:
A simple regex tester. The user enters:
1) A regular expression pattern
2) Optional flags (i=IGNORECASE, m=MULTILINE, s=DOTALL)
3) A test string

The program prints each match and any captured groups.
"""

import re

pattern = input("Enter a regex pattern: ")

flags_input = input("Enter flags (i/m/s) or press Enter for none: ").strip().lower()
flags = 0

if "i" in flags_input:
  flags |= re.IGNORECASE
if "m" in flags_input:
  flags |= re.MULTILINE
if "s" in flags_input:
  flags |= re.DOTALL

text = input("Enter the test string: ")

try:
  compiled = re.compile(pattern, flags)
except re.error as error:
  print("Regex error:", error)
  raise SystemExit

matches = list(compiled.finditer(text))

if len(matches) == 0:
  print("No matches found.")
else:
  print(f"Matches found: {len(matches)}\n")
  for idx, match in enumerate(matches, start=1):
    print(f"Match {idx}: '{match.group(0)}'  at {match.start()}..{match.end()}")

    if match.groups():
      for g_idx, g_val in enumerate(match.groups(), start=1):
        print(f"  Group {g_idx}: '{g_val}'")

    print()
```
