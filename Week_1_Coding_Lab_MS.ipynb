```python
"""
1. Positive or Negative: Write a Python program that asks the user for
a number and determines whether this number is positive or negative.
"""

def is_positive_or_negative():
  """
  Description: Ask the user for a number and determine if it's positive
  or negative.

  Inputs: An integer from the user.

  Output: Prints whether the number is positive, negative, or zero.
  """
  number = int(input("Enter a random number: \n"))

  if number < 0:
    print(f"{number} is negative!")
  elif number > 0:
    print(f"{number} is positive!")
  else:
    print("0 is neither positive or negative.")

is_positive_or_negative() # Entering 3 will result in "3 is positive!"
```

```python
"""
2. Even or Odd: Write a Python program that asks the user for a number and
determines whether this number is even or odd.
"""

def is_even_or_odd():
  """
  Description: Ask the user for a number and determine if it's even or odd.

  Inputs: An integer from the user.

  Output: Prints whether the number is even or odd.
  """
  number = int(input("Enter a random number: \n"))

  if number % 2 == 0:
    print(f"{number} is an even number!")
  else:
    print(f"{number} is an odd number!")

is_even_or_odd() # Entering 2 will result in "2 is an even number"
```

```python
"""
3. Sum of Numbers Write a Python program that calculates the sum of all
numbers from 1 to a number n provided by the user."""

def sum_to_n():
  """
  Description: Calculate the sum of all numbers from 1 to n.

  Inputs: An integer n from the user.

  Output: Prints the sum from 1 to n.
  """
  number = int(input("Enter a random number: \n"))

  sum = 0
  while number > 0:
    sum = sum + number
    number = number - 1

  print(f"Sum of 1 to {number} is {sum}")

sum_to_n() # Entering 3 will output "Sum of 1 to 0 is 6"
```

```python
"""
4. Multiplication Table Write a Python program that prints the multiplication table
 (from 1 to 10) of a number provided by the user.
"""

def print_multiplication_table():
  """
  Description: Print the multiplication table (1 to 10) for a user-provided
  number.

  Inputs: An integer from the user.

  Output: Prints formatted multiplication results from 1 through 10.
  """
  number = int(input("Enter a random number: \n"))

  max = 1

  while max < 11:
    current_product = number * max
    print(f"{number} * {max} = {current_product}")
    max = max + 1

print_multiplication_table() # Entering 3 will output the multiplication table for 3

```

```python
"""
5. Prime Numbers: Write a Python program that asks the user for a
number n and prints all prime numbers less than the entered number
"""

import math

def is_prime(num):
  """
  Description: Determine if a given integer is a prime number.

  Inputs: The number to test.

  Output: True if candidate is prime, False otherwise.

  Prime number definition: A prime number is an integer greater than 1 that
  has no divisors other than 1 and itself.
  """
  if num <= 1: # Sanity check even though it gets checked in the get_prime_below_number function as well
    return False
  if num == 2:
    return True
  if num % 2 == 0:
    return False

  for i in range(2, int(math.sqrt(num)) + 1):
    if (num % i) == 0:
      return False
  return True

def get_prime_below_number():
  """
  Description: Takes user input and calls the is_prime() function to test
  all numbers below that number for primes.
  """
  number = int(input("Enter a random number: \n"))
  primes = []
  for num in range(2, number + 1):
    if is_prime(num):
      primes.append(num)
  return primes

get_prime_below_number() # Entering 7 will output [2, 3, 5, 7]
```
