```python
"""List Comprehensions and Filters: Write a Python program that uses a list
comprehension to generate a list of the squares of the numbers from 1 to 10.
Then, use the filter function to keep only the odd numbers squares."""

"""
Description: This program uses a list comprehension to generate a list of
the squares of the numbers from 1 to 10. Then, it uses the filter function
to keep only the odd numbers squares.

Input: Predefined: 1 to 10

Output:
[1, 9, 25, 49, 81]
"""

even_squares = [x**2 for x in range(1,11) if x%2 == 1]
print(even_squares)
```

```python
"""Generators and Tuples: Write a Python generator function that generates
Fibonacci sequence. The function should return a tuple with the index and
the Fibonacci number for each call. Use the generator to print the first 10
Fibonacci numbers."""

"""
Description: A generator function that generates the Fibonacci sequence.
The function yields a tuple with the index and the Fibonacci number for each call.

Input: None

Output: A tuple with the index and the Fibonacci number for each call.

Example output:
(0, 0)
(1, 1)
(2, 1)
(3, 2)
"""

def fibonacci_generator():
  a, b = 0, 1
  index = 0
  while True:
    yield index, a
    a, b = b, a + b
    index +=1

fibonacci = fibonacci_generator()
for _ in range(10):
  index, number = next(fibonacci)
  print((index, number))
```

```python
"""Maps and Reduces: Write a Python program that uses the map function to
calculate the cube of each number in a list. Then, use the reduce function
from the functools module to find the sum of the cubes."""

"""
Description: This program uses the map function to calculate the cube of each
number in a list. Then, the reduce function from the functools module is used
to find the sum of the cubes.

Input:
- The program takes a list of numbers as input from the user.

Output:
- The program prints the sum of the cubes

Example:
- Enter a list of numbers seperated by space.
  Example: 1 2 3
  Output: 36
"""

from functools import reduce

def cube(x):
  return x**3

def sum_cubes(numbers):
  cubes = [x for x in map(cube, numbers)]
  return reduce(lambda x, y: x + y, cubes)

numbers = [1,2,3]
print(sum_cubes(numbers))
```

```python
"""Word Frequency with Dictionaries: Write a Python program that takes a
sentence as input from the user and uses a dictionary to store the frequency
of each word in the sentence. The keys in the dictionary should be the words,
and the values should be the frequencies."""

"""
Description: This program takes a sentence as input from the user and uses a
dictionary to store the frequency of each word in the sentence.

Input:
- The program takes a sentence as input from the user.

Output:
- The program prints the frequency of each word in the sentence

Example:
- Enter a sentence: This is a test.
- {'This': 1, 'is': 1, 'a': 1, 'test.': 1}
"""

def word_frequency():
  sentence = input("Enter a sentence: ")
  words = sentence.split()
  word_frequency = {}

  for word in words:
    if word in word_frequency:
      word_frequency[word] += 1
    else:
      word_frequency[word] = 1
  return word_frequency

print(word_frequency())
```

```python
"""Unique Elements with Sets: Write a Python program that takes a list of
numbers as input from the user and uses a set to find and print the unique
numbers in the list."""

"""
Description: This program takes a list of numbers as input from the user and
uses a set to find and print the unique numbers in the list

Input:
- The program takes a list of numbers as input from the user.

Output:
- The program prints the unique numbers in the list

Example:
- Enter a list of numbers seperated by space.
  Example: 1 2 2
  Output: {1, 2}
"""

def unique_elements():
  numbers = input("Enter a list of numbers seperated by space.\nExample: 1 2 2: ")
  numbers = [int(x) for x in numbers.split()]
  return set(numbers)

unique_elements()
```
